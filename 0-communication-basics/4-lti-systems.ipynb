## 2. 概念讲解

### 2.1 线性时不变系统定义

**线性系统（Linear System）**

线性系统满足**叠加原理**，即：

1. **可加性**：若输入x₁(t)→输出y₁(t)，输入x₂(t)→输出y₂(t)，则输入x₁(t)+x₂(t)→输出y₁(t)+y₂(t)

2. **齐次性**：若输入x(t)→输出y(t)，则输入ax(t)→输出ay(t)（a为常数）

综合表达：
$$T\{a x_1(t) + b x_2(t)\} = a T\{x_1(t)\} + b T\{x_2(t)\}$$

**时不变系统（Time-Invariant System）**

时不变系统的特性不随时间变化：
$$若\; x(t) \rightarrow y(t), 则\; x(t-t_0) \rightarrow y(t-t_0)$$

即：输入延迟t₀，输出也延迟t₀，系统的输入输出关系保持不变。

| 性质 | 定义 | 物理意义 |
|------|------|----------|
| 线性 | 叠加原理成立 | 系统对多个输入的响应等于各输入响应的叠加 |
| 时不变 | 系统参数不随时间变化 | 今天和昨天的系统行为完全相同 |
| **LTI** | 同时满足线性和时不变 | 可用卷积表示输入输出关系 |

### 2.2 连续卷积

对于连续时间LTI系统，**卷积积分**表示输入信号与系统冲激响应的关系：

$$y(t) = x(t) * h(t) = \int_{-\infty}^{\infty} x(\tau) h(t-\tau) d\tau$$

其中：
- $x(t)$：输入信号
- $h(t)$：系统的**冲激响应**（对单位冲激信号δ(t)的响应）
- $y(t)$：输出信号

**卷积的物理意义**：
- 任意输入可以分解为一系列加权冲激信号的叠加
- 每个冲激信号在τ时刻的强度为x(τ)dτ
- 系统对每个冲激的响应为x(τ)dτ × h(t-τ)
- 总响应是所有冲激响应的积分（叠加）

**卷积运算律**：
| 定律 | 表达式 | 说明 |
|------|--------|------|
| 交换律 | $x(t)*h(t) = h(t)*x(t)$ | 输入与系统可以交换顺序 |
| 结合律 | $(x*h_1)*h_2 = x*(h_1*h_2)$ | 级联系统等效为一个系统 |
| 分配律 | $x*(h_1+h_2) = x*h_1 + x*h_2$ | 并联系统响应可叠加 |

### 2.3 离散卷积

对于离散时间LTI系统，**卷积和**表示输入序列与系统冲激响应的关系：

$$y[n] = x[n] * h[n] = \sum_{k=-\infty}^{\infty} x[k] \cdot h[n-k]$$

其中：
- $x[n]$：输入序列
- $h[n]$：系统的**单位脉冲响应**
- $y[n]$：输出序列

**计算步骤**：
1. 将h[k]翻转（关于k=0对称）得到h[-k]
2. 将h[-k]平移n个单位得到h[n-k]
3. 将x[k]与h[n-k]相乘并求和
4. 对所有n重复上述过程

| 连续系统 | 离散系统 |
|-----------|----------|
| 卷积积分 ∫ | 卷积和 Σ |
| 连续时间t | 离散索引n |
| h(t): 冲激响应 | h[n]: 脉冲响应 |
| 积分区间(-∞,∞) | 求和区间(-∞,∞) |

### 2.4 冲激响应与系统特性

**冲激响应h(t)**是LTI系统最重要的特性：

**定义**：系统对单位冲激信号δ(t)的响应
$$h(t) = T\{\delta(t)\}$$

**冲激响应完全刻画了LTI系统的特性**：
- 已知h(t)，可通过卷积计算任意输入的输出
- 无法找到其他更简单的系统表示方法

**由冲激响应判断系统特性**：

| 系统特性 | h(t)/h[n]的特征 | 判断方法 |
|----------|------------------|----------|
| **因果性** | h(t)=0, t<0 | 输出仅取决于当前及过去输入 |
| **稳定性** | ∫|h(t)|dt < ∞ | 有界输入产生有界输出(BIBO) |
| **记忆性** | h(t)≠kδ(t) | 系统当前输出与历史输入有关 |

**因果系统**：
$$h(t) = 0, \quad t < 0$$

**稳定系统**（BIBO稳定）：
$$\int_{-\infty}^{\infty} |h(t)| dt < \infty \quad \text{或} \quad \sum_{n=-\infty}^{\infty} |h[n]| < \infty$$

### 2.5 频率响应

频率响应是系统对正弦信号的响应特性，是LTI系统分析的核心工具。

**连续时间系统的频率响应**：
$$H(j\omega) = \int_{-\infty}^{\infty} h(t) e^{-j\omega t} dt$$

即：频率响应是冲激响应的傅里叶变换。

**离散时间系统的频率响应**：
$$H(e^{j\omega}) = \sum_{n=-\infty}^{\infty} h[n] e^{-j\omega n}$$

**幅度响应与相位响应**：
$$H(j\omega) = |H(j\omega)| e^{j\phi(\omega)}$$

其中：
- $|H(j\omega)|$：**幅度响应**，表示系统对不同频率的增益
- $\phi(\omega) = \arg\{H(j\omega)\}$：**相位响应**，表示系统对不同频率的相移

**正弦稳态响应**：
若输入 $x(t) = A\cos(\omega_0 t + \theta)$，则输出
$$y(t) = A|H(j\omega_0)|\cos(\omega_0 t + \theta + \phi(\omega_0))$$

即：幅度响应改变正弦波的幅度，相位响应改变正弦波的相位。

### 2.6 卷积定理

**卷积定理**是时域与频域分析的核心桥梁：

**时域卷积 ↔ 频域乘积**
$$若\; y(t) = x(t) * h(t), 则\; Y(\omega) = X(\omega) \cdot H(\omega)$$

**频域卷积 ↔ 时域乘积**
$$若\; y(t) = x(t) \cdot h(t), 则\; Y(\omega) = \frac{1}{2\pi} X(\omega) * H(\omega)$$

| 时域操作 | 频域结果 | 实际应用 |
|----------|----------|----------|
| 卷积 | 乘积 | 系统响应 = 输入频谱 × 频率响应 |
| 乘积 | 卷积 | 调制 = 频谱与载频频谱的卷积 |

**卷积定理的应用**：
1. **系统分析**：计算输出频谱 Y(ω) = X(ω)·H(ω)
2. **系统设计**：通过设计 H(ω) 实现想要的滤波效果
3. **计算加速**：时域卷积可通过FFT转换为频域乘积计算

## 3. Python演示

下面我们使用Python来演示LTI系统的相关概念。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

print("LTI系统演示环境已准备就绪")
```

### 3.1 离散卷积的直观演示

```python
def demo_discrete_convolution():
    """
    演示离散卷积的计算过程
    展示翻转、平移、相乘、求和的完整步骤

    参数说明：
    - x: 输入序列
    - h: 脉冲响应序列
    """
    
    # 示例1：简单序列的卷积
    # 输入序列：x = [1, 2, 1]（相当于 x[0]=1, x[1]=2, x[2]=1）
    x = np.array([1, 2, 1])
    # 脉冲响应：h = [1, 1, 1]（矩形滤波器）
    h = np.array([1, 1, 1])
    
    # 使用numpy计算卷积（验证用）
    y_numpy = np.convolve(x, h, mode='full')
    
    # 手动计算卷积，展示过程
    n_x = len(x)
    n_h = len(h)
    n_y = n_x + n_h - 1  # 卷积结果长度
    
    print("离散卷积计算演示")
    print("=" * 60)
    print(f"输入序列 x: {x}")
    print(f"脉冲响应 h: {h}")
    print(f"卷积结果 y = x * h: {y_numpy}")
    print("-" * 60)
    
    # 可视化卷积过程
    fig, axes = plt.subplots(4, 1, figsize=(12, 12))
    
    # 绘制原始序列
    n = np.arange(len(x))
    axes[0].stem(n, x, linefmt='b-', markerfmt='bo', basefmt=' ')
    axes[0].set_ylabel('x[n]', fontsize=11)
    axes[0].set_title('输入序列 x[n]', fontsize=12)
    axes[0].grid(True, alpha=0.3)
    axes[0].set_ylim(-0.5, 3)
    
    k = np.arange(len(h))
    axes[1].stem(k, h, linefmt='g-', markerfmt='go', basefmt=' ')
    axes[1].set_ylabel('h[n]', fontsize=11)
    axes[1].set_title('脉冲响应 h[n]', fontsize=12)
    axes[1].grid(True, alpha=0.3)
    axes[1].set_ylim(-0.5, 3)
    
    # 绘制翻转和平移过程
    n_h_flip = np.arange(len(h)-1, -1, -1)  # 翻转后的索引
    h_flip = h[::-1]  # 翻转脉冲响应
    
    # 绘制翻转后的h
    axes[2].stem(n_h_flip, h_flip, linefmt='r-', markerfmt='ro', basefmt=' ')
    axes[2].set_ylabel('h[-k]', fontsize=11)
    axes[2].set_title('翻转后的脉冲响应 h[-k]', fontsize=12)
    axes[2].grid(True, alpha=0.3)
    axes[2].set_ylim(-0.5, 3)
    
    # 绘制卷积结果
    n_y = np.arange(len(y_numpy))
    axes[3].stem(n_y, y_numpy, linefmt='m-', markerfmt='mo', basefmt=' ')
    axes[3].set_ylabel('y[n]', fontsize=11)
    axes[3].set_xlabel('n', fontsize=11)
    axes[3].set_title('卷积结果 y[n] = x[n] * h[n]', fontsize=12)
    axes[3].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 打印计算过程
    print("\n卷积计算过程（手动计算）：")
    print("-" * 60)
    print("y[n] = Σ x[k] · h[n-k]")
    print("-" * 60)
    for n_val in range(len(y_numpy)):
        # 计算y[n_val]的各个乘积项
        terms = []
        for k_val in range(len(x)):
            idx_h = n_val - k_val
            if 0 <= idx_h < len(h):
                term = x[k_val] * h[idx_h]
                terms.append(f"x[{k_val}]·h[{idx_h}]={term}")
        y_val = sum(x[k_val] * h[n_val - k_val] for k_val in range(len(x)) 
                     if 0 <= n_val - k_val < len(h))
        if terms:
            print(f"y[{n_val}] = " + " + ".join(terms) + f" = {y_val}")
    print("-" * 60)
    
    # 示例2：矩形脉冲与指数衰减序列的卷积
    print("\n" + "=" * 60)
    print("示例2：矩形脉冲与指数衰减序列的卷积")
    print("=" * 60)
    
    # 矩形脉冲（持续5个采样点）
    x2 = np.ones(5)
    # 指数衰减脉冲响应
    h2 = np.array([1, 0.5, 0.25, 0.125, 0.0625])
    
    y2 = np.convolve(x2, h2, mode='full')
    
    print(f"矩形脉冲 x: {x2}")
    print(f"指数响应 h: {h2}")
    print(f"卷积结果 y: {y2}")
    
    # 可视化
    fig, axes = plt.subplots(3, 1, figsize=(12, 9))
    
    n_x2 = np.arange(len(x2))
    axes[0].stem(n_x2, x2, linefmt='b-', markerfmt='bo', basefmt=' ')
    axes[0].set_ylabel('x[n]', fontsize=11)
    axes[0].set_title('矩形脉冲输入', fontsize=12)
    axes[0].grid(True, alpha=0.3)
    axes[0].set_ylim(-0.1, 1.5)
    
    n_h2 = np.arange(len(h2))
    axes[1].stem(n_h2, h2, linefmt='g-', markerfmt='go', basefmt=' ')
    axes[1].set_ylabel('h[n]', fontsize=11)
    axes[1].set_title('指数衰减脉冲响应', fontsize=12)
    axes[1].grid(True, alpha=0.3)
    axes[1].set_ylim(-0.1, 1.5)
    
    n_y2 = np.arange(len(y2))
    axes[2].stem(n_y2, y2, linefmt='r-', markerfmt='ro', basefmt=' ')
    axes[2].set_ylabel('y[n]', fontsize=11)
    axes[2].set_xlabel('n', fontsize=11)
    axes[2].set_title('卷积结果（三角形叠加效应）', fontsize=12)
    axes[2].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n卷积分析结论：")
    print("1. 卷积将输入信号的形状与脉冲响应的形状进行叠加")
    print("2. 当脉冲响应是矩形时，输出是输入的'滑动平均'")
    print("3. 当脉冲响应是指数衰减时，输出会有拖尾效应")
    
    return y_numpy

y_result = demo_discrete_convolution()
```

### 3.2 连续卷积的离散近似

```python
def demo_continuous_convolution():
    """
    演示连续时间卷积的离散近似计算
    使用高采样率来近似连续卷积结果

    参数说明：
    - x: 连续输入信号（矩形脉冲）
    - h: 连续冲激响应（指数衰减）
    - fs: 采样率
    """
    
    fs = 1000  # 高采样率，用于近似连续信号
    T = 2.0   # 信号持续时间
    t = np.linspace(0, T, int(fs * T))
    
    # 创建矩形脉冲输入信号 x(t)
    # 脉冲宽度 0.5秒，幅度 1
    tau = 0.5
    x = np.where((t >= 0) & (t <= tau), 1.0, 0.0)
    
    # 创建指数衰减冲激响应 h(t)
    # h(t) = e^(-t) * u(t)，其中u(t)是单位阶跃函数
    alpha = 2.0
    h = np.exp(-alpha * t) * np.where(t >= 0, 1.0, 0.0)
    
    # 使用scipy的convolve进行离散卷积（近似连续卷积）
    # 由于采样率高，离散卷积结果接近连续卷积
    dt = 1 / fs
    y_scipy = signal.convolve(x, h, mode='full') * dt  # 乘以dt近似积分
    
    # 创建对应的时间轴
    t_y = np.linspace(0, 2*T, len(y_scipy))
    
    # 理论计算（分段分析）
    # 对于 y(t) = x(t) * h(t)，其中x是矩形，h是指数衰减
    # 当 0 <= t <= tau: y(t) = (1 - e^(-alpha*t)) / alpha
    # 当 tau < t: y(t) = (e^(-alpha*(t-tau)) - e^(-alpha*t)) / alpha
    
    y_theory = np.zeros_like(t_y)
    for i, ti in enumerate(t_y):
        if ti <= tau:
            y_theory[i] = (1 - np.exp(-alpha * ti)) / alpha
        else:
            y_theory[i] = (np.exp(-alpha * (ti - tau)) - np.exp(-alpha * ti)) / alpha
    
    # 可视化
    fig, axes = plt.subplots(3, 1, figsize=(12, 10))
    
    # 输入信号
    axes[0].plot(t, x, 'b-', linewidth=2)
    axes[0].set_ylabel('x(t)', fontsize=11)
    axes[0].set_title('连续时间输入信号：矩形脉冲', fontsize=12)
    axes[0].grid(True, alpha=0.3)
    axes[0].set_xlim(0, 2)
    axes[0].set_ylim(-0.1, 1.5)
    axes[0].axhline(y=0, color='k', linewidth=0.5)
    axes[0].axvline(x=0, color='k', linewidth=0.5)
    
    # 冲激响应
    axes[1].plot(t, h, 'g-', linewidth=2)
    axes[1].set_ylabel('h(t)', fontsize=11)
    axes[1].set_title('连续时间冲激响应：指数衰减 h(t) = e^(-2t)u(t)', fontsize=12)
    axes[1].grid(True, alpha=0.3)
    axes[1].set_xlim(0, 2)
    axes[1].set_ylim(-0.1, 1.5)
    axes[1].axhline(y=0, color='k', linewidth=0.5)
    axes[1].axvline(x=0, color='k', linewidth=0.5)
    
    # 卷积结果
    axes[2].plot(t_y, y_scipy[:len(t_y)], 'r-', linewidth=1.5, 
                  label='scipy卷积结果', alpha=0.8)
    axes[2].plot(t_y, y_theory, 'b--', linewidth=2, 
                  label='理论计算结果')
    axes[2].set_ylabel('y(t)', fontsize=11)
    axes[2].set_xlabel('时间 (s)', fontsize=11)
    axes[2].set_title('连续时间卷积结果：y(t) = x(t) * h(t)', fontsize=12)
    axes[2].legend(loc='upper right')
    axes[2].grid(True, alpha=0.3)
    axes[2].set_xlim(0, 3)
    axes[2].set_ylim(-0.1, 0.8)
    axes[2].axhline(y=0, color='k', linewidth=0.5)
    axes[2].axvline(x=0, color='k', linewidth=0.5)
    
    plt.tight_layout()
    plt.show()
    
    # 分析卷积结果的特点
    print("\n连续卷积分析结论：")
    print("=" * 60)
    print("1. 输入是宽度为0.5s的矩形脉冲，幅度为1")
    print("2. 冲激响应是指数衰减：h(t) = e^(-2t)u(t)")
    print("3. 卷积结果的特点：")
    print("   - 在t=0到t=τ期间：输出从0逐渐增加")
    print("   - 在t=τ之后：输出达到峰值后逐渐衰减")
    print("4. 物理意义：")
    print("   - 指数衰减系统的'记忆'效应：输出不仅取决于当前输入")
    print("   - 还取决于之前的输入（历史积累效应）")
    print("=" * 60)
    
    return t_y, y_scipy[:len(t_y)]

t_conv, y_conv = demo_continuous_convolution()
```

### 3.3 不同系统对同一输入的响应对比

### 3.4 频率响应可视化

```python
def demo_frequency_response():
    """
    演示典型LTI系统的频率响应
    展示幅度响应和相位响应的物理意义

    参数说明：
    - 系统类型：低通、高通、带通
    """
    
    # 定义频率范围
    f = np.linspace(0, 100, 1000)  # 0-100Hz
    omega = 2 * np.pi * f
    
    # 创建不同的系统及其频率响应
    systems = {}
    
    # 1. RC低通滤波器: H(jw) = 1/(1 + jwRC)
    # 截止频率 fc = 1/(2*pi*RC) = 10Hz
    fc_low = 10  # 截止频率
    tau_low = 1 / (2 * np.pi * fc_low)
    H_low = 1 / (1 + 1j * omega * tau_low)
    systems['RC低通滤波器'] = {'H': H_low, 'f': f}
    
    # 2. RC高通滤波器: H(jw) = jwRC/(1 + jwRC)
    fc_high = 15  # 截止频率
    tau_high = 1 / (2 * np.pi * fc_high)
    H_high = (1j * omega * tau_high) / (1 + 1j * omega * tau_high)
    systems['RC高通滤波器'] = {'H': H_high, 'f': f}
    
    # 3. 二阶带通滤波器（谐振电路）
    f0 = 30  # 中心频率
    Q = 5    # 品质因数
    H_bp = (1j * omega / (2 * np.pi * f0)) / (1 + (1j * omega / (2 * np.pi * f0 * Q)) - (omega / (2 * np.pi * f0))**2)
    # 简化模型
    H_bp = 1 / (1 + 1j * Q * (f/f0 - f0/f) / f0 * f0)
    H_bp = np.abs(1 / (1 + 1j * Q * (f/f0 - f0/f)))
    H_bp = np.where(f > 0, H_bp, 0)
    systems['带通滤波器'] = {'H': H_bp, 'f': f, 'simple': True}
    
    # 创建图形
    fig, axes = plt.subplots(len(systems), 2, figsize=(14, 12))
    
    colors = ['blue', 'red', 'green']
    
    for idx, (name, sys_data) in enumerate(systems.items()):
        H = sys_data['H']
        
        # 幅度响应
        ax_mag = axes[idx, 0]
        
        if sys_data.get('simple'):
            ax_mag.plot(f, np.abs(H), color=colors[idx], linewidth=2)
        else:
            ax_mag.plot(f, np.abs(H), color=colors[idx], linewidth=2)
        ax_mag.set_ylabel('|H(f)|', fontsize=11)
        ax_mag.set_title(f'{name} - 幅度响应', fontsize=12)
        ax_mag.grid(True, alpha=0.3)
        ax_mag.set_xlim(0, 100)
        ax_mag.set_ylim(0, 1.2)
        
        # 相位响应
        ax_phase = axes[idx, 1]
        
        if not sys_data.get('simple'):
            phase = np.angle(H, deg=True)
            ax_phase.plot(f, phase, color=colors[idx], linewidth=2)
            ax_phase.set_ylabel('∠H(f) (度)', fontsize=11)
            ax_phase.set_xlabel('频率 (Hz)', fontsize=11)
        else:
            ax_phase.text(0.5, 0.5, '相位响应
(带通特性)', 
                          transform=ax_phase.transAxes, 
                          ha='center', va='center', fontsize=12)
            ax_phase.set_ylabel('∠H(f) (度)', fontsize=11)
            ax_phase.set_xlabel('频率 (Hz)', fontsize=11)
        
        ax_phase.set_title(f'{name} - 相位响应', fontsize=12)
        ax_phase.grid(True, alpha=0.3)
        ax_phase.set_xlim(0, 100)
    
    # 添加最后一个轴标签
    axes[-1, 0].set_xlabel('频率 (Hz)', fontsize=11)
    if not sys_data.get('simple'):
        axes[-1, 1].set_xlabel('频率 (Hz)', fontsize=11)
    
    plt.tight_layout()
    plt.show()
    
    # 打印物理意义解释
    print("\n频率响应物理意义：")
    print("=" * 60)
    print("1. 幅度响应 |H(f)|：")
    print("   - 表示系统对不同频率信号的增益")
    print("   - |H(f)|=1：该频率成分无衰减通过")
    print("   - |H(f)|=0：该频率成分被完全抑制")
    print("   - 0<|H(f)|<1：该频率成分部分通过，有衰减")
    print("")
    print("2. 相位响应 ∠H(f)：")
    print("   - 表示系统对不同频率信号的相移")
    print("   - 相位延迟 = ∠H(f)/ω = ∠H(f)/(2πf) 秒")
    print("   - 线性相位：所有频率延迟相同，信号不失真")
    print("   - 非线性相位：不同频率延迟不同，信号失真")
    print("")
    print("3. 实际滤波器的特性：")
    print("   - 低通滤波器：让低频通过，抑制高频")
    print("   - 高通滤波器：让高频通过，抑制低频")
    print("   - 带通滤波器：让特定频段通过，抑制其他频率")
    print("=" * 60)
    
    # 演示：正弦信号通过系统的响应
    print("\n正弦信号通过系统示例：")
    print("-" * 60)
    
    # 输入信号
    fs = 1000
    t = np.linspace(0, 0.5, int(fs * 0.5))
    f_in = 5  # 5Hz正弦信号
    x = np.sin(2 * np.pi * f_in * t)
    
    # 通过RC低通滤波器
    # 计算系统在5Hz的频率响应
    H_at_f = 1 / (1 + 1j * 2 * np.pi * f_in * tau_low)
    magnitude = np.abs(H_at_f)
    phase_shift = np.angle(H_at_f)
    
    # 输出信号
    y = magnitude * np.sin(2 * np.pi * f_in * t + phase_shift)
    
    print(f"输入信号: {f_in}Hz正弦波")
    print(f"系统在{f_in}Hz的增益: {magnitude:.4f}")
    print(f"系统在{f_in}Hz的相移: {np.degrees(phase_shift):.2f}度")
    print(f"输出信号幅度: {magnitude:.4f} (输入幅度为1)")
    
    # 绘图
    fig, ax = plt.subplots(figsize=(12, 5))
    ax.plot(t, x, 'b-', linewidth=1.5, label='输入信号', alpha=0.7)
    ax.plot(t, y, 'r-', linewidth=1.5, label='输出信号')
    ax.set_xlabel('时间 (s)', fontsize=11)
    ax.set_ylabel('幅度', fontsize=11)
    ax.set_title(f'5Hz正弦信号通过RC低通滤波器（fc=10Hz）', fontsize=12)
    ax.legend(loc='upper right')
    ax.grid(True, alpha=0.3)
    ax.set_xlim(0, 0.2)
    plt.tight_layout()
    plt.show()
    
    print("\n观察结论：")
    print(f"- 由于5Hz < 截止频率10Hz，信号大部分通过（增益≈{magnitude:.2f}）")
    print("- 输出信号相对于输入有相位延迟")
    print("- 幅度略有衰减（因为5Hz接近截止频率）")
    
    return

demo_frequency_response()
```

### 3.5 卷积定理验证

```python
def demo_convolution_theorem():
    """
    验证卷积定理：时域卷积 = 频域乘积
    展示两种计算方法的结果一致性

    参数说明：
    - x: 输入信号
    - h: 系统冲激响应
    """
    
    fs = 1000
    T = 1.0
    t = np.linspace(0, T, int(fs * T))
    
    # 创建输入信号：两个正弦波的叠加
    x = np.sin(2*np.pi*5*t) + 0.5*np.sin(2*np.pi*12*t)
    
    # 创建系统冲激响应：指数衰减
    h = np.exp(-10*t) * (t >= 0)
    
    # 方法1：直接在时域进行卷积
    y_time = signal.convolve(x, h, mode='same') / fs
    
    # 方法2：使用卷积定理 - 在频域计算
    # 计算各信号的频谱
    X = np.fft.fft(x)
    H = np.fft.fft(h)
    
    # 频域乘积
    Y_freq = X * H
    
    # 逆变换回时域
    y_freq = np.fft.ifft(Y_freq).real / fs
    
    # 方法3：使用FFT加速的卷积
    n = len(x) + len(h) - 1
    X_fft = np.fft.fft(x, n)
    H_fft = np.fft.fft(h, n)
    Y_fft = X_fft * H_fft
    y_fft = np.fft.ifft(Y_fft).real / fs
    
    # 可视化结果对比
    fig, axes = plt.subplots(3, 2, figsize=(14, 12))
    
    # 绘制输入信号和频谱
    axes[0, 0].plot(t, x, 'b-', linewidth=1.5)
    axes[0, 0].set_ylabel('x(t)', fontsize=11)
    axes[0, 0].set_title('输入信号（5Hz + 12Hz）', fontsize=12)
    axes[0, 0].grid(True, alpha=0.3)
    
    freq_x = np.fft.fftfreq(len(x), 1/fs)
    axes[0, 1].plot(freq_x, np.abs(np.fft.fftshift(X))/len(X), 'b-', linewidth=1.5)
    axes[0, 1].set_ylabel('|X(f)|', fontsize=11)
    axes[0, 1].set_title('输入信号频谱', fontsize=12)
    axes[0, 1].set_xlim(-30, 30)
    axes[0, 1].grid(True, alpha=0.3)
    
    # 绘制冲激响应和频谱
    axes[1, 0].plot(t, h, 'g-', linewidth=1.5)
    axes[1, 0].set_ylabel('h(t)', fontsize=11)
    axes[1, 0].set_title('冲激响应（指数衰减）', fontsize=12)
    axes[1, 0].grid(True, alpha=0.3)
    
    freq_h = np.fft.fftfreq(len(h), 1/fs)
    H_shift = np.fft.fftshift(H)
    axes[1, 1].plot(np.fft.fftshift(freq_h), np.abs(H_shift)/len(h), 'g-', linewidth=1.5)
    axes[1, 1].set_ylabel('|H(f)|', fontsize=11)
    axes[1, 1].set_title('系统频率响应', fontsize=12)
    axes[1, 1].set_xlim(-50, 50)
    axes[1, 1].grid(True, alpha=0.3)
    
    # 绘制输出结果对比
    t_y = np.linspace(0, T, len(y_time))
    axes[2, 0].plot(t_y, y_time, 'r-', linewidth=1.5, label='时域卷积', alpha=0.8)
    t_y2 = np.linspace(0, T, len(y_freq))
    axes[2, 0].plot(t_y2, y_freq, 'b--', linewidth=1.5, label='频域乘积（numpy）', alpha=0.8)
    axes[2, 0].set_ylabel('y(t)', fontsize=11)
    axes[2, 0].set_xlabel('时间 (s)', fontsize=11)
    axes[2, 0].set_title('卷积结果对比：时域卷积 vs 频域乘积', fontsize=12)
    axes[2, 0].legend(loc='upper right')
    axes[2, 0].grid(True, alpha=0.3)
    
    # 绘制输出频谱
    Y_shift = np.fft.fftshift(Y_freq)
    freq_y = np.fft.fftfreq(len(Y_freq), 1/fs)
    axes[2, 1].plot(np.fft.fftshift(freq_y), np.abs(Y_shift)/len(Y_shift), 'r-', linewidth=1.5)
    axes[2, 1].set_ylabel('|Y(f)|', fontsize=11)
    axes[2, 1].set_xlabel('频率 (Hz)', fontsize=11)
    axes[2, 1].set_title('输出信号频谱 Y(f) = X(f) × H(f)', fontsize=12)
    axes[2, 1].set_xlim(-30, 30)
    axes[2, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 计算两种方法的误差
    error = np.abs(y_time - y_freq[:len(y_time)])
    max_error = np.max(error)
    mean_error = np.mean(error)
    
    print("\n卷积定理验证结果：")
    print("=" * 60)
    print(f"时域卷积与频域乘积的最大误差: {max_error:.2e}")
    print(f"时域卷积与频域乘积的平均误差: {mean_error:.2e}")
    print("-" * 60)
    print("结论：两种方法的结果高度一致，验证了卷积定理")
    print("      y(t) = x(t) * h(t)  ⟺  Y(f) = X(f) · H(f)")
    print("=" * 60)
    
    # FFT加速卷积的优势
    import time
    
    # 时域卷积的计算复杂度: O(N×M)
    # FFT卷积的计算复杂度: O(NlogN)
    print("\n计算效率对比（大规模数据时）：")
    print("-" * 60)
    print("时域卷积: O(N × M)，每次移位都需要N×M次乘法")
    print("FFT卷积:  O(N log N)，只需几次FFT和逆FFT")
    print("当N很大时，FFT卷积远快于直接时域卷积")
    
    return

demo_convolution_theorem()
```

## 4. 原理解析

### 4.1 LTI系统分析的核心思想

```python
def demo_lti_core_concept():
    """
    演示LTI系统分析的核心思想：
    任意输入 → 分解为基本信号的叠加 → 基本信号响应 → 叠加
    """
    
    print("LTI系统分析的核心思想")
    print("=" * 60)
    print("\n为什么LTI系统如此重要？")
    print("-" * 60)
    print("1. 线性性 + 时不变性 → 输入输出关系可完全用卷积描述")
    print("2. 冲激响应 h(t) 完全刻画系统特性")
    print("3. 一旦知道 h(t)，任何输入的输出都可以计算")
    print("\n" + "=" * 60)
    
    print("\nLTI系统分析的三种方法：")
    print("-" * 60)
    print("\n【方法1】直接卷积法")
    print("   y(t) = ∫ x(τ)h(t-τ) dτ")
    print("   优点：概念清晰，适用于任何输入")
    print("   缺点：积分计算可能复杂")
    
    print("\n【方法2】频域分析法（最常用！）")
    print("   X(f) = FT{x(t)}")
    print("   H(f) = FT{h(t)}")
    print("   Y(f) = X(f) × H(f)")
    print("   y(t) = IFT{Y(f)}")
    print("   优点：卷积变乘积，复数运算简单")
    print("   缺点：需要理解频域表示")
    
    print("\n【方法3】状态空间法")
    print("   用于多输入多输出系统")
    print("   现代控制系统的基础")
    
    print("\n" + "=" * 60)
    
    # 创建一个综合示例
    fs = 1000
    t = np.linspace(0, 2, int(fs * 2))
    
    # 复杂输入信号
    x = (np.sin(2*np.pi*2*t) +           # 2Hz低频
         0.7*np.sin(2*np.pi*10*t) +      # 10Hz中频
         0.3*np.sin(2*np.pi*25*t)) *     # 25Hz高频
        np.exp(-t/2)                     # 指数包络
    
    # 低通系统
    fc = 15  # 截止频率
    tau = 1 / (2*np.pi*fc)
    h_lp = (1/tau) * np.exp(-t/tau) * (t >= 0)
    
    # 高通系统
    h_hp = np.exp(-t/0.05) * (t >= 0)
    h_hp = np.gradient(h_hp, t)  # 近似微分
    
    # 计算输出
    y_lp = signal.convolve(x, h_lp, mode='same') / fs
    y_hp = signal.convolve(x, h_hp, mode='same') / fs
    
    # 可视化
    fig, axes = plt.subplots(3, 1, figsize=(12, 10))
    
    axes[0].plot(t, x, 'b-', linewidth=1.5)
    axes[0].set_ylabel('x(t)', fontsize=11)
    axes[0].set_title('复杂输入信号（含多种频率成分）', fontsize=12)
    axes[0].grid(True, alpha=0.3)
    
    axes[1].plot(t, y_lp, 'r-', linewidth=1.5)
    axes[1].set_ylabel('y_lp(t)', fontsize=11)
    axes[1].set_title('通过低通滤波器（fc=15Hz）- 高频成分被抑制', fontsize=12)
    axes[1].grid(True, alpha=0.3)
    
    axes[2].plot(t, y_hp, 'g-', linewidth=1.5)
    axes[2].set_ylabel('y_hp(t)', fontsize=11)
    axes[2].set_xlabel('时间 (s)', fontsize=11)
    axes[2].set_title('通过高通滤波器 - 低频成分被抑制', fontsize=12)
    axes[2].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n观察结论：")
    print("- 低通滤波器输出：平滑了输入，保留了低频波动")
    print("- 高通滤波器输出：提取了输入的突变部分")
    print("- 同一个输入通过不同系统，得到完全不同的输出")
    print("- 这正是滤波器的核心作用：选择性地处理信号")
    
    return

demo_lti_core_concept()
```

### 4.2 因果性与稳定性的判定

```python
def demo_causality_stability():
    """
    演示因果性和稳定性的判定方法
    
    因果性：h(t) = 0, t < 0
    稳定性：∫|h(t)|dt < ∞
    """
    
    print("因果性与稳定性判定")
    print("=" * 60)
    
    fs = 1000
    t = np.linspace(-1, 2, int(fs * 3))
    
    # 定义几个典型系统
    systems = {
        '因果低通': {
            'h': np.where(t >= 0, 10*np.exp(-10*t), 0),
            'desc': 'h(t) = e^(-10t)u(t), t≥0 时非零'
        },
        '非因果低通': {
            'h': 10*np.exp(-10*np.abs(t)),
            'desc': 'h(t) = e^(-10|t|), t<0 时也有值'
        },
        '不稳定系统': {
            'h': np.where(t >= 0, np.exp(2*t), 0),
            'desc': 'h(t) = e^(2t)u(t), 积分不收敛'
        }
    }
    
    fig, axes = plt.subplots(3, 2, figsize=(14, 12))
    
    results = []
    
    for idx, (name, sys_info) in enumerate(systems.items()):
        h = sys_info['h']
        
        # 判定因果性
        # 因果系统：h(t) = 0 for t < 0
        h_negative = h[t < 0]
        is_causal = np.allclose(h_negative, np.zeros_like(h_negative))
        
        # 判定稳定性
        # 稳定系统：∫|h(t)|dt < ∞
        # 近似计算积分
        integral = np.sum(np.abs(h)) * (t[1] - t[0])
        is_stable = integral < np.inf and not np.isnan(integral)
        
        # 对于增长型指数，检查数值稳定性
        if np.max(np.abs(h)) > 1e10:
            is_stable = False
        
        results.append({
            'name': name,
            'desc': sys_info['desc'],
            'causal': is_causal,
            'stable': is_stable
        })
        
        # 绘制冲激响应
        axes[idx, 0].plot(t, h, 'b-', linewidth=2)
        axes[idx, 0].axvline(x=0, color='r', linestyle='--', alpha=0.5)
        axes[idx, 0].axhline(y=0, color='k', linewidth=0.5)
        axes[idx, 0].set_ylabel('h(t)', fontsize=11)
        axes[idx, 0].set_title(f'{name} - 冲激响应', fontsize=12)
        axes[idx, 0].grid(True, alpha=0.3)
        
        # 标注因果性
        causal_text = "因果" if is_causal else "非因果"
        stable_text = "稳定" if is_stable else "不稳定"
        axes[idx, 0].text(0.02, 0.95, f'{causal_text}, {stable_text}', 
                         transform=axes[idx, 0].transAxes, fontsize=12,
                         verticalalignment='top',
                         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
        
        # 绘制频率响应（幅度）
        H = np.fft.fft(h, len(t))
        freq = np.fft.fftfreq(len(t), t[1]-t[0])
        H_shift = np.fft.fftshift(H)
        freq_shift = np.fft.fftshift(freq)
        
        # 只显示正值频率（因果系统的频率响应）
        pos_mask = freq_shift >= 0
        axes[idx, 1].plot(freq_shift[pos_mask], np.abs(H_shift[pos_mask]), 'r-', linewidth=2)
        axes[idx, 1].set_ylabel('|H(f)|', fontsize=11)
        axes[idx, 1].set_xlabel('频率 (Hz)', fontsize=11)
        axes[idx, 1].set_title(f'{name} - 频率响应', fontsize=12)
        axes[idx, 1].grid(True, alpha=0.3)
        axes[idx, 1].set_xlim(0, 50)
    
    plt.tight_layout()
    plt.show()
    
    # 打印判定结果
    print("\n系统特性判定结果：")
    print("=" * 70)
    print("{:<20} | {:<35} | 因果性 | 稳定性".format("系统名称", "冲激响应描述"))
    print("-" * 70)
    for r in results:
        causal_str = "是" if r['causal'] else "否"
        stable_str = "是" if r['stable'] else "否"
        print("{:<20} | {:<35} |   {:^5} |   {:^5}".format(
            r['name'], r['desc'], causal_str, stable_str))
    print("=" * 70)
    
    print("\n判定方法总结：")
    print("-" * 60)
    print("【因果性判定】")
    print("   条件：h(t) = 0, 对于所有 t < 0")
    print("   物理意义：输出不依赖于未来输入")
    print("   实际系统通常是因果的（非实时系统可以是非因果）")
    print("\n【稳定性判定】")
    print("   条件：∫|h(t)|dt < ∞ 或 ∑|h[n]| < ∞")
    print("   BIBO稳定：有界输入产生有界输出")
    print("   不稳定系统：微小扰动可能导致输出无限增长")
    print("-" * 60)
    
    return

demo_causality_stability()
```

## 总结

本notebook介绍了LTI系统的核心概念：

- **LTI系统**：线性时不变系统，同时满足叠加原理和时不变性
- **卷积**：LTI系统输入输出关系的基本表达 y(t) = x(t)*h(t)
- **冲激响应**：系统对单位冲激信号的响应，完全刻画系统特性
- **因果性**：h(t)=0, t<0，输出仅取决于当前及过去输入
- **稳定性**：∫|h(t)|dt<∞，BIBO稳定，有界输入产生有界输出
- **频率响应**：H(jω)是h(t)的傅里叶变换，表示系统对不同频率的响应
- **卷积定理**：时域卷积等于频域乘积，连接了时域和频域分析

LTI系统理论是信号与系统、通信原理、控制系统等课程的核心基础。
