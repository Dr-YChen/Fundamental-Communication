## 2. 概念讲解

### 2.1 傅里叶级数：周期信号的频域表示

**傅里叶级数（Fourier Series）**用于表示周期信号。任何周期为$T_0$的周期信号可以分解为一系列正弦和余弦波的叠加：

**三角形式傅里叶级数**
$$x(t) = a_0 + \sum_{n=1}^{\infty} \left[ a_n \cos(2\pi n f_0 t) + b_n \sin(2\pi n f_0 t) \right]$$

其中：
- $f_0 = 1/T_0$：基频（fundamental frequency）
- $n f_0$：谐波频率（harmonic frequencies）
- 系数 $a_n, b_n$ 决定了各频率成分的幅度

**指数形式傅里叶级数**
$$x(t) = \sum_{n=-\infty}^{\infty} c_n e^{j2\pi n f_0 t}$$

其中 $c_n$ 是复数系数，包含幅度和相位信息。

| 特性 | 说明 |
|------|------|
| 周期性 | 频谱是离散的，仅在基频的整数倍处有成分 |
| 离散性 | 频率成分是离散的谱线 |
| 收敛性 | 满足狄利克雷条件时，级数收敛 |
| 正交性 | 正弦余弦函数在周期内正交 |

**收敛条件（狄利克雷条件）**：
1. 在一个周期内，信号绝对可积
2. 在一个周期内，信号只有有限个极值点
3. 在一个周期内，信号只有有限个不连续点

### 2.2 傅里叶变换：非周期信号的频谱

**傅里叶变换（Fourier Transform）**用于分析非周期信号。当周期信号的周期趋向无穷大时，傅里叶级数变成傅里叶变换：

**傅里叶变换**
$$X(f) = \int_{-\infty}^{\infty} x(t) e^{-j2\pi ft} dt$$

**逆傅里叶变换**
$$x(t) = \int_{-\infty}^{\infty} X(f) e^{j2\pi ft} df$$

| 特性 | 傅里叶级数 | 傅里叶变换 |
|------|-----------|------------|
| 信号类型 | 周期信号 | 非周期信号 |
| 频谱类型 | 离散频谱 | 连续频谱 |
| 频率变量 | $n\omega_0$ (离散) | $\omega$ (连续) |
| 变换结果 | 复数系数 $c_n$ | 连续函数 $X(f)$ |

### 2.3 幅度谱 vs 相位谱

傅里叶变换的结果 $X(f)$ 是复数函数，可以表示为：

$$X(f) = |X(f)| e^{j\phi(f)}$$

**幅度谱（Magnitude Spectrum）**
- 定义：$|X(f)|$，表示信号中各个频率成分的幅度
- 物理意义：反映信号在不同频率的能量分布
- 偶函数：$|X(-f)| = |X(f)|$

**相位谱（Phase Spectrum）**
- 定义：$\phi(f) = \arg\{X(f)\}$，表示各频率成分的相位关系
- 物理意义：决定信号的时域波形结构
- 奇函数：$\phi(-f) = -\phi(f)$

**重要理解**：幅度谱决定信号包含哪些频率，相位谱决定这些频率如何组合成时域波形

### 2.4 能量信号 vs 功率信号

**能量信号**
- 总能量有限：$E = \int_{-\infty}^{\infty} |x(t)|^2 dt < \infty$
- 平均功率为0：$P = \lim_{T\to\infty} \frac{1}{T}\int_{-T/2}^{T/2} |x(t)|^2 dt = 0$
- 例子：脉冲信号、有限持续时间的信号
- 频域表示：能量谱密度 $G(f) = |X(f)|^2$

**功率信号**
- 总能量无限，但平均功率有限：$0 < P < \infty$
- 周期信号都是功率信号
- 例子：无限持续的正弦波、白噪声
- 频域表示：功率谱密度 $S(f)$

| 类型 | 能量 | 功率 | 例子 |
|------|------|------|------|
| 能量信号 | 有限 | 0 | 矩形脉冲、高斯脉冲 |
| 功率信号 | 无限 | 有限 | 正弦波、周期信号、白噪声 |

### 2.5 帕塞瓦尔定理（Parseval's Theorem）

帕塞瓦尔定理表明：**信号的总能量在时域和频域是守恒的**

**连续时间信号**
$$\int_{-\infty}^{\infty} |x(t)|^2 dt = \int_{-\infty}^{\infty} |X(f)|^2 df$$

其中 $|X(f)|^2$ 称为**能量谱密度**，表示信号能量在频率上的分布。

**物理意义**：
- 时域中计算的总能量等于频域中计算的总能量
- 是通信系统中功率预算和带宽分析的理论基础
- 说明频域分析不会丢失任何信息

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

print("傅里叶分析演示环境已准备就绪")
```

### 3.1 周期信号的傅里叶级数分解

```python
def demo_fourier_series():
    """
    演示周期信号的傅里叶级数分解
    使用方波作为例子，展示如何将周期信号分解为各次谐波
    
    参数说明：
    - T0: 信号周期 (秒)
    - f0: 基频 (Hz)
    - N_harmonics: 谐波数量
    """
    
    # 参数设置
    T0 = 1.0      # 周期 1秒
    f0 = 1.0      # 基频 1Hz
    t = np.linspace(-2, 2, 2000)  # 时间轴，覆盖2个周期
    
    # 创建方波信号（周期信号示例）
    # 方波：占空比50%，幅度在+1和-1之间切换
    square_wave = signal.square(2 * np.pi * f0 * t, duty=0.5)
    
    # 计算傅里叶级数系数（对于对称方波，只有奇次谐波）
    # 方波的傅里叶级数系数：an = 0 (n为偶数), an = 4/(n*pi) (n为奇数)
    N_harmonics = 15  # 使用的谐波数量
    
    # 重构信号：从傅里叶级数
    reconstructed = np.zeros_like(t)
    harmonic_amplitudes = []
    
    for n in range(1, N_harmonics + 1, 2):  # 只有奇次谐波
        # 方波的傅里叶系数：4/(n*pi)
        an = 4 / (n * np.pi)
        harmonic_amplitudes.append((n, an))
        # 叠加正弦波（方波是奇函数，只有正弦项）
        reconstructed += an * np.sin(2 * np.pi * n * f0 * t)
    
    # 可视化
    fig, axes = plt.subplots(3, 1, figsize=(12, 10))
    
    # 图1：原始方波 vs 重构信号
    ax1 = axes[0]
    ax1.plot(t, square_wave, 'b-', linewidth=2, label='原始方波', alpha=0.7)
    ax1.plot(t, reconstructed, 'r--', linewidth=1.5, label=f'傅里叶级数重构 (N={N_harmonics})')
    ax1.set_xlabel('时间 (s)', fontsize=10)
    ax1.set_ylabel('幅度', fontsize=10)
    ax1.set_title('周期方波信号的傅里叶级数分解与重构', fontsize=12)
    ax1.legend(loc='upper right')
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(-1.5, 1.5)
    ax1.axhline(y=0, color='k', linewidth=0.5)
    
    # 图2：各次谐波分量
    ax2 = axes[1]
    colors = plt.cm.rainbow(np.linspace(0, 1, len(harmonic_amplitudes)))
    for i, (n, an) in enumerate(harmonic_amplitudes[:5]):  # 只显示前5个谐波
        harmonic = an * np.sin(2 * np.pi * n * f0 * t)
        ax2.plot(t, harmonic, color=colors[i], linewidth=1, alpha=0.7,
                 label=f'第{n}次谐波 (f={n}Hz, 幅度={an:.3f})')
    ax2.set_xlabel('时间 (s)', fontsize=10)
    ax2.set_ylabel('幅度', fontsize=10)
    ax2.set_title('前5个正弦谐波分量', fontsize=12)
    ax2.legend(loc='upper right', fontsize=8)
    ax2.grid(True, alpha=0.3)
    ax2.set_ylim(-1.5, 1.5)
    
    # 图3：频谱（幅度谱）
    ax3 = axes[2]
    n_values = [h[0] for h in harmonic_amplitudes]
    a_values = [h[1] for h in harmonic_amplitudes]
    ax3.stem(n_values, a_values, basefmt=' ', linefmt='b-', markerfmt='bo')
    ax3.set_xlabel('谐波次数 n', fontsize=10)
    ax3.set_ylabel('幅度', fontsize=10)
    ax3.set_title('幅度谱 (各次谐波的幅度)', fontsize=12)
    ax3.set_xticks(n_values)
    ax3.set_xticklabels([f'{n}\n({n}Hz)' for n in n_values])
    ax3.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 打印谐波信息
    print("方波的傅里叶级数分析：")
    print("=" * 50)
    print("谐波次数 | 频率(Hz) | 理论幅度 | 实际幅度")
    print("-" * 50)
    for n, an in harmonic_amplitudes:
        actual = 4 / (n * np.pi)
        print(f"   {n:2d}    |   {n:3d}   |  {an:.4f}   |  {actual:.4f}")
    print("=" * 50)
    print("\n注意：方波只有奇次谐波，幅度随谐波次数增加而减小")
    
    return harmonic_amplitudes

harmonic_data = demo_fourier_series()
```

```python
print("\n傅里叶级数分析结论：")
print("1. 方波包含无限多的奇次谐波（1, 3, 5, 7, ...）")
print("2. 各谐波幅度为 4/(n*pi)，高次谐波幅度逐渐减小")
print("3. 谐波数量越多，重构信号越接近原始方波")
print("4. 使用15次谐波已能较好地逼近方波")
print("\n提示：实际工程中，谐波衰减特性影响信号带宽设计")
```

### 3.2 非周期信号的频谱分析

```python
def demo_fourier_transform():
    """
    演示非周期信号的傅里叶变换
    使用矩形脉冲和高斯脉冲作为例子，展示连续频谱
    
    参数说明：
    - fs: 采样率 (Hz)
    - T: 信号持续时间 (秒)
    """
    
    # 参数设置
    fs = 1000   # 采样率 1000Hz
    T = 2.0     # 信号持续时间 2秒
    t = np.linspace(-T/2, T/2, int(fs * T))
    
    # 创建矩形脉冲（非周期能量信号）
    tau = 0.2   # 脉冲宽度 0.2秒
    rect_pulse = np.where(np.abs(t) <= tau/2, 1.0, 0.0)
    
    # 创建高斯脉冲
    sigma = 0.1  # 高斯标准差
    gaussian_pulse = np.exp(-t**2 / (2 * sigma**2))
    
    # 计算傅里叶变换（使用FFT近似）
    # 注意：FFT计算的是离散频谱，需要正确标定
    def compute_fft(x, fs):
        """计算信号的FFT并返回正确的频率轴"""
        N = len(x)
        X = np.fft.fftshift(np.fft.fft(x))  # 移到中心
        freq = np.fft.fftfreq(N, 1/fs)  # 频率轴
        freq = np.fft.fftshift(freq)  # 移到中心
        # 标定幅度（除以采样点数）
        X = X / N
        return freq, X
    
    freq_rect, X_rect = compute_fft(rect_pulse, fs)
    freq_gauss, X_gauss = compute_fft(gaussian_pulse, fs)
    
    # 理论频谱对比
    # 矩形脉冲的频谱：sinc函数
    freq_theory = np.linspace(-50, 50, 1000)
    sinc_spectrum = tau * np.sinc(freq_theory * tau / np.pi)
    
    # 可视化
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # 图1：矩形脉冲时域
    ax1 = axes[0, 0]
    ax1.plot(t, rect_pulse, 'b-', linewidth=2)
    ax1.set_xlabel('时间 (s)', fontsize=10)
    ax1.set_ylabel('幅度', fontsize=10)
    ax1.set_title('矩形脉冲时域波形', fontsize=12)
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(-0.1, 1.3)
    ax1.axvline(x=-tau/2, color='r', linestyle='--', alpha=0.5)
    ax1.axvline(x=tau/2, color='r', linestyle='--', alpha=0.5)
    ax1.text(0, 1.1, f'宽度 τ={tau}s', ha='center', fontsize=10)
    
    # 图2：矩形脉冲频域（幅度谱）
    ax2 = axes[0, 1]
    ax2.plot(freq_theory, np.abs(sinc_spectrum), 'b-', linewidth=2, label='理论sinc函数')
    ax2.plot(freq_rect, np.abs(X_rect), 'r--', linewidth=1.5, label='FFT计算结果')
    ax2.set_xlabel('频率 (Hz)', fontsize=10)
    ax2.set_ylabel('|X(f)|', fontsize=10)
    ax2.set_title('矩形脉冲幅度谱 (连续频谱)', fontsize=12)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    ax2.set_xlim(-50, 50)
    
    # 图3：高斯脉冲时域
    ax3 = axes[1, 0]
    ax3.plot(t, gaussian_pulse, 'g-', linewidth=2)
    ax3.set_xlabel('时间 (s)', fontsize=10)
    ax3.set_ylabel('幅度', fontsize=10)
    ax3.set_title('高斯脉冲时域波形', fontsize=12)
    ax3.grid(True, alpha=0.3)
    ax3.set_ylim(-0.1, 1.1)
    
    # 图4：高斯脉冲频域（也是高斯形状）
    ax4 = axes[1, 1]
    ax4.plot(freq_gauss, np.abs(X_gauss), 'g-', linewidth=2)
    ax4.set_xlabel('频率 (Hz)', fontsize=10)
    ax4.set_ylabel('|X(f)|', fontsize=10)
    ax4.set_title('高斯脉冲幅度谱 (仍是高斯形状)', fontsize=12)
    ax4.grid(True, alpha=0.3)
    ax4.set_xlim(-50, 50)
    
    plt.tight_layout()
    plt.show()
    
    # 分析：主瓣宽度与脉冲宽度的关系
    print("\n频谱分析结论：")
    print("=" * 60)
    print("1. 矩形脉冲的频谱是sinc函数，主瓣宽度与脉冲宽度成反比")
    print("2. 脉冲越窄，频谱越宽（时域窄→频域宽，这是时频对偶性）")
    print("3. 高斯脉冲的频谱也是高斯形状，特殊情况：时频域形状相同")
    print("4. 非周期信号的频谱是连续的，不同于周期信号的离散谱线")
    print("=" * 60)
    
    # 计算第一零点位置（理论值）
    null_freq = 1 / tau  # 第一零点频率
    print(f"\n矩形脉冲参数：")
    print(f"  - 脉冲宽度 τ = {tau}s")
    print(f"  - 第一零点频率 = 1/τ = {null_freq}Hz")
    print(f"  - 有效带宽（近似） = 2/τ = {2*null_freq}Hz")
    
    return

demo_fourier_transform()
```

### 3.3 多个频率成分的信号合成

```python
def demo_signal_synthesis():
    """
    演示多个频率成分如何构成复杂信号
    展示：不同幅度和相位的正弦波叠加可以产生各种波形
    
    参数说明：
    - fs: 采样率 (Hz)
    - T: 信号持续时间 (秒)
    """
    
    fs = 1000  # 采样率 1000Hz
    T = 1.0   # 持续时间 1秒
    t = np.linspace(0, T, int(fs * T))
    
    # 定义三个不同频率、幅度、相位的正弦波
    components = [
        {'freq': 5, 'amp': 1.0, 'phase': 0},      # 5Hz, 幅度1, 初相0
        {'freq': 10, 'amp': 0.5, 'phase': np.pi/4}, # 10Hz, 幅度0.5, 相位π/4
        {'freq': 15, 'amp': 0.3, 'phase': np.pi/2}, # 15Hz, 幅度0.3, 相位π/2
    ]
    
    # 生成各分量信号
    signals = []
    for comp in components:
        sig = comp['amp'] * np.sin(2 * np.pi * comp['freq'] * t + comp['phase'])
        signals.append(sig)
    
    # 合成信号
    composite = sum(signals)
    
    # 计算合成信号的频谱
    N = len(composite)
    X = np.fft.fftshift(np.fft.fft(composite))
    freq = np.fft.fftfreq(N, 1/fs)
    freq = np.fft.fftshift(freq)
    X = X / N
    
    # 可视化
    fig, axes = plt.subplots(3, 2, figsize=(14, 12))
    
    # 上排：各分量时域波形
    colors = ['blue', 'green', 'orange']
    for i, (comp, sig) in enumerate(zip(components, signals)):
        ax = axes[0, i]
        ax.plot(t[:500], sig[:500], color=colors[i], linewidth=1.5)
        ax.set_title(f'分量 {i+1}: {comp["freq"]}Hz, 幅度={comp["amp"]}, 相位={comp["phase"]/np.pi:.2f}π', 
                    fontsize=11)
        ax.set_xlabel('时间 (s)', fontsize=9)
        ax.set_ylabel('幅度', fontsize=9)
        ax.grid(True, alpha=0.3)
        ax.set_ylim(-1.8, 1.8)
    
    # 中图：合成信号时域
    ax_comp = axes[1, 0]
    ax_comp.plot(t[:500], composite[:500], 'r-', linewidth=1.5)
    ax_comp.set_title('合成信号时域波形', fontsize=12)
    ax_comp.set_xlabel('时间 (s)', fontsize=10)
    ax_comp.set_ylabel('幅度', fontsize=10)
    ax_comp.grid(True, alpha=0.3)
    ax_comp.set_ylim(-2, 2)
    
    # 中右：合成信号频谱（幅度谱）
    ax_spec = axes[1, 1]
    ax_spec.stem(freq, np.abs(X), basefmt=' ', linefmt='r-', markerfmt='ro')
    ax_spec.set_title('合成信号幅度谱', fontsize=12)
    ax_spec.set_xlabel('频率 (Hz)', fontsize=10)
    ax_spec.set_ylabel('|X(f)|', fontsize=10)
    ax_spec.set_xlim(-25, 25)
    ax_spec.grid(True, alpha=0.3)
    
    # 标记各频率成分的位置
    for comp in components:
        ax_spec.axvline(x=comp['freq'], color='b', linestyle='--', alpha=0.3)
        ax_spec.axvline(x=-comp['freq'], color='b', linestyle='--', alpha=0.3)
    
    # 下排：相位谱
    ax_phase = axes[2, 0]
    ax_phase.stem(freq, np.angle(X), basefmt=' ', linefmt='g-', markerfmt='go')
    ax_phase.set_title('合成信号相位谱', fontsize=12)
    ax_phase.set_xlabel('频率 (Hz)', fontsize=10)
    ax_phase.set_ylabel('相位 (rad)', fontsize=10)
    ax_phase.set_xlim(-25, 25)
    ax_phase.grid(True, alpha=0.3)
    
    # 下右：帕塞瓦尔定理验证
    ax_pv = axes[2, 1]
    
    # 计算时域能量
    E_time = np.sum(composite**2) / fs
    # 计算频域能量
    E_freq = np.sum(np.abs(X)**2) * (fs/N)
    
    # 绘制能量对比图
    bar_labels = ['时域计算', '频域计算']
    bar_values = [E_time, E_freq]
    ax_pv.bar(bar_labels, bar_values, color=['blue', 'orange'], alpha=0.7)
    ax_pv.set_title(f'帕塞瓦尔定理验证 (能量={E_time:.4f})', fontsize=12)
    ax_pv.set_ylabel('能量', fontsize=10)
    ax_pv.grid(True, alpha=0.3, axis='y')
    
    for i, v in enumerate(bar_values):
        ax_pv.text(i, v + 0.01, f'{v:.6f}', ha='center', fontsize=10)
    
    plt.tight_layout()
    plt.show()
    
    # 打印分析结果
    print("\n多频率成分信号分析：")
    print("=" * 60)
    print("各频率分量：")
    for i, comp in enumerate(components):
        print(f"  分量{i+1}: f={comp['freq']}Hz, 幅度={comp['amp']}, 相位={comp['phase']/np.pi:.2f}π")
    print("-" * 60)
    print(f"时域能量: {E_time:.6f}")
    print(f"频域能量: {E_freq:.6f}")
    print(f"相对误差: {abs(E_time-E_freq)/E_time*100:.4f}%")
    print("=" * 60)
    print("结论：帕塞瓦尔定理验证，时域和频域能量守恒")
    
    return

demo_signal_synthesis()
```

## 4. 原理解析

### 4.1 时频对偶性

时频对偶性是傅里叶分析的核心概念：

| 时域特性 | 频域特性 | 说明 |
|----------|----------|------|
| 窄脉冲 | 宽频谱 | 脉冲越窄，包含的频率成分越丰富 |
| 宽脉冲 | 窄频谱 | 脉冲越宽，频谱越集中 |
| 周期信号 | 离散频谱 | 周期 → 频谱离散化 |
| 非周期信号 | 连续频谱 | 非周期 → 频谱连续化 |

**时宽-带宽积**：对于有限持续时间的信号，其时宽和带宽的乘积近似为常数：

$$\Delta t \cdot \Delta f \approx constant$$

这意味着我们无法同时获得无限的时间分辨率和频率分辨率。

```python
def demo_time_frequency_duality():
    """
    演示时频对偶性
    展示脉冲宽度与频谱宽度的关系
    """
    
    fs = 1000
    T = 2.0
    t = np.linspace(-T/2, T/2, int(fs * T))
    
    # 不同宽度的矩形脉冲
    widths = [0.05, 0.1, 0.2, 0.4]  # 脉冲宽度
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    axes = axes.flatten()
    
    for i, tau in enumerate(widths):
        # 创建矩形脉冲
        rect = np.where(np.abs(t) <= tau/2, 1.0, 0.0)
        
        # 计算频谱
        X = np.fft.fftshift(np.fft.fft(rect))
        freq = np.fft.fftfreq(len(rect), 1/fs)
        freq = np.fft.fftshift(freq)
        X = X / len(X)
        
        # 绘制时域和频域
        ax = axes[i]
        
        # 时域图（蓝色）
        ax2 = ax.twinx()
        ax.plot(t, rect, 'b-', linewidth=2, label=f'时域 (τ={tau}s)')
        ax2.plot(freq, np.abs(X), 'r-', linewidth=1.5, label='频域 |X(f)|')
        
        ax.set_xlabel('时间(s) / 频率(Hz)', fontsize=10)
        ax.set_ylabel('时域幅度', color='blue', fontsize=10)
        ax2.set_ylabel('频域幅度', color='red', fontsize=10)
        ax.set_title(f'脉冲宽度 τ={tau}s → 带宽 ≈ {2/tau:.0f}Hz', fontsize=12)
        
        # 设置x轴范围以显示主要部分
        ax.set_xlim(-1, 1)
        ax2.set_xlim(-100, 100)
        
        ax.grid(True, alpha=0.3)
        
        # 添加图例
        ax.legend(loc='upper left')
        ax2.legend(loc='upper right')
    
    plt.tight_layout()
    plt.show()
    
    print("\n时频对偶性观察：")
    print("=" * 50)
    print("脉冲宽度 τ | 第一零点频率 | 时宽×带宽 |")
    print("-" * 50)
    for tau in widths:
        null_freq = 1 / tau
        tw_product = tau * null_freq
        print(f"   {tau:.2f}s   |    {null_freq:.1f}Hz    |   {tw_product:.2f}   ")
    print("=" * 50)
    print("结论：τ×(1/τ) ≈ 1，时宽-带宽积近似为常数")
    print("      脉冲越窄，频谱越宽；脉冲越宽，频谱越窄")
    
demo_time_frequency_duality()
```

### 4.2 快速傅里叶变换（FFT）简介

**FFT（Fast Fourier Transform）**是计算DFT的高效算法，将计算复杂度从$O(N^2)$降低到$O(N \log N)$。

**FFT的基本思想**：
1. **分治法**：将大点数DFT分解为小点数DFT
2. **蝶形运算**：通过反复的加减运算组合子问题结果
3. **基数-2 FFT**：要求信号长度为$2^n$

**FFT参数理解**：
| 参数 | 含义 | 说明 |
|------|------|------|
| $N$ | FFT点数 | 必须是$2^n$，影响频率分辨率 |
| $f_s$ | 采样率 | 决定可分析的频率范围 |
| $\Delta f = f_s/N$ | 频率分辨率 | 取决于FFT点数和采样率 |
| $f_{max} = f_s/2$ | 最大分析频率 | 遵循Nyquist采样定理 |

**频率分辨率与信号长度的关系**：
$$\Delta f = \frac{1}{N \cdot T_s} = \frac{f_s}{N}$$

其中$T_s = 1/f_s$是采样周期。

```python
def demo_fft_parameters():
    """
    演示FFT参数对频谱分析的影响
    展示采样率、FFT点数与频率分辨率的关系
    """
    
    # 测试信号：包含10Hz和12Hz两个频率成分
    f1, f2 = 10, 12  # 两个接近的频率
    
    fs = 100  # 采样率
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # 情况1：低采样率，不够长的观测窗口
    N1 = 64   # 64点FFT
    t1 = np.arange(N1) / fs
    x1 = np.sin(2*np.pi*f1*t1) + np.sin(2*np.pi*f2*t1)
    
    X1 = np.fft.fftshift(np.fft.fft(x1))
    freq1 = np.fft.fftfreq(N1, 1/fs)
    freq1 = np.fft.fftshift(freq1)
    
    ax1 = axes[0, 0]
    ax1.stem(freq1, np.abs(X1)/N1, basefmt=' ')
    ax1.set_title(f'情况1: fs={fs}Hz, N={N1}点, Δf={fs/N1:.2f}Hz\n(无法分辨10Hz和12Hz)', fontsize=11)
    ax1.set_xlabel('频率 (Hz)', fontsize=10)
    ax1.set_ylabel('|X(f)|', fontsize=10)
    ax1.set_xlim(-20, 20)
    ax1.grid(True, alpha=0.3)
    
    # 情况2：相同采样率，更长的观测窗口
    N2 = 256  # 256点FFT
    t2 = np.arange(N2) / fs
    x2 = np.sin(2*np.pi*f1*t2) + np.sin(2*np.pi*f2*t2)
    
    X2 = np.fft.fftshift(np.fft.fft(x2))
    freq2 = np.fft.fftfreq(N2, 1/fs)
    freq2 = np.fft.fftshift(freq2)
    
    ax2 = axes[0, 1]
    ax2.stem(freq2, np.abs(X2)/N2, basefmt=' ')
    ax2.set_title(f'情况2: fs={fs}Hz, N={N2}点, Δf={fs/N2:.2f}Hz\n(可以分辨10Hz和12Hz)', fontsize=11)
    ax2.set_xlabel('频率 (Hz)', fontsize=10)
    ax2.set_ylabel('|X(f)|', fontsize=10)
    ax2.set_xlim(-20, 20)
    ax2.grid(True, alpha=0.3)
    
    # 情况3：更高采样率
    fs3 = 200
    N3 = 256
    t3 = np.arange(N3) / fs3
    x3 = np.sin(2*np.pi*f1*t3) + np.sin(2*np.pi*f2*t3)
    
    X3 = np.fft.fftshift(np.fft.fft(x3))
    freq3 = np.fft.fftfreq(N3, 1/fs3)
    freq3 = np.fft.fftshift(freq3)
    
    ax3 = axes[1, 0]
    ax3.stem(freq3, np.abs(X3)/N3, basefmt=' ')
    ax3.set_title(f'情况3: fs={fs3}Hz, N={N3}点, Δf={fs3/N3:.2f}Hz\n(频率范围扩大，分辨率不变)', fontsize=11)
    ax3.set_xlabel('频率 (Hz)', fontsize=10)
    ax3.set_ylabel('|X(f)|', fontsize=10)
    ax3.set_xlim(-20, 20)
    ax3.grid(True, alpha=0.3)
    
    # 情况4：总结表格
    ax4 = axes[1, 1]
    ax4.axis('off')
    
    # 创建总结信息
    summary = """
    FFT参数对比总结
    ================

    情况 | fs(Hz) | N(点) | Δf(Hz) | 可分辨？
    -----|--------|-------|--------|--------
     1   |  100   |   64  |  1.56  |  否(Δf>2Hz)
     2   |  100   |  256  |  0.39  |  是(Δf<2Hz)
     3   |  200   |  256  |  0.78  |  是(Δf<2Hz)
    
    关键结论：
    • 频率分辨率 Δf = fs/N
    • 提高分辨率的方法：
      1. 增加N（观测时间更长）
      2. 降低fs（但需满足Nyquist条件）
    • 两个频率的最小可分辨间隔 ≈ 2×Δf
    """
    ax4.text(0.1, 0.9, summary, transform=ax4.transAxes, fontsize=12,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.show()
    
    print("\nFFT参数设计建议：")
    print("1. 根据信号最高频率确定采样率：fs ≥ 2f_max")
    print("2. 根据频率分辨率要求确定FFT点数：N = fs/Δf")
    print("3. 如果N不够大，延长观测时间（加零 padding 或真实延长）")
    print("4. 频率分辨率取决于观测窗口长度：T = N/fs = 1/Δf")
    
demo_fft_parameters()
```
