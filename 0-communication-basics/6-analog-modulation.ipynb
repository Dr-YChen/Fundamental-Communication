## 2. 概念讲解

### 2.1 为什么要调制？

调制是将信息信号（基带信号）附加到高频载波上的过程。在无线通信中，调制是必不可少的环节，原因如下：

| 调制的原因 | 说明 | 重要性 |
|------------|------|--------|
| **频谱效率** | 多个调制信号可以占据不同频段，实现频分复用 | 充分利用有限频谱资源 |
| **天线尺寸** | 天线长度与信号波长成正比，频率越高天线越短 | 使便携设备成为可能 |
| **无线传输** | 基带信号无法有效辐射（需要天线与波长可比） | 实现真正的无线通信 |
| **抗干扰** | 高频调制信号对噪声和干扰更具鲁棒性 | 提高通信可靠性 |
| **信道匹配** | 匹配信道特性，优化传输效率 | 适应不同传输环境 |

**天线长度与频率的关系**：
对于有效辐射，天线长度通常需要达到波长的1/4到1/2：

$λ = \frac{c}{f}$

其中 c ≈ 3×10⁸ m/s 是光速。例如，300Hz的音频信号需要天线长度约250km，而3GHz的微波信号只需约2.5cm的天线。

### 2.2 幅度调制（AM/DSB/SSB）

幅度调制是最基本的调制方式，通过改变载波的幅度来传输信息。

**标准AM（常规幅度调制）**：

调幅信号表达式：

$$s(t) = [1 + m \cdot x(t)] \cdot \cos(2\pi f_c t)$$

其中：
- $x(t)$：基带调制信号（假设已归一化，|x(t)| ≤ 1）
- $m$：调制度（0 < m ≤ 1），表示调制深度
- $f_c$：载波频率

**调制度m的意义**：
- m = 0：未调制，只有纯载波
- m = 1：满调制，载波幅度在0~2之间变化
- m > 1：过调制，会导致信号失真

**DSB-SC（双边带抑制载波）**：

$$s(t) = x(t) \cdot \cos(2\pi f_c t)$$

与标准AM相比，DSB-SC去掉了载波分量，因此功率效率更高，但解调需要同步载波。

**SSB（单边带）**：

只传输上边带或下边带之一，进一步节省带宽：

$$s_{USB}(t) = x(t) \cdot \cos(2\pi f_c t) - \hat{x}(t) \cdot \sin(2\pi f_c t)$$

其中 $\hat{x}(t)$ 是 $x(t)$ 的希尔伯特变换。

### 2.3 频率调制（FM）

频率调制通过改变载波的瞬时频率来传输信息。

**瞬时频率的概念**：

对于信号 $s(t) = A \cos(\phi(t))$，瞬时频率定义为相位对时间的导数除以2π：

$$f_i(t) = \frac{1}{2\pi} \frac{d\phi(t)}{dt}$$

**调频信号的表达式**：

$$s(t) = A_c \cos[2\pi f_c t + 2\pi k_f \int_0^t x(\tau) d\tau]$$

其中 $k_f$ 是频率灵敏度（Hz/V）。

**调频的关键参数**：

| 参数 | 公式 | 说明 |
|------|------|------|
| **调制指数 β** | $\beta = \frac{k_f \cdot A_m}{f_m} = \frac{\Delta f}{f_m}$ | 峰值频偏与调制频率的比值 |
| **峰值频偏 Δf** | $\Delta f = k_f \cdot A_m$ | 最大频率偏移 |
| ** Carson 带宽** | $B_T ≈ 2(\Delta f + f_m) = 2f_m(\beta + 1)$ | 近似包含98%功率的带宽 |

**FM的特点**：
- 幅度恒定，抗幅度噪声能力强
- 频谱扩展，带宽利用率低于AM
- 具有天然的抗干扰能力（噪声主要影响幅度）
- 需要较宽的传输带宽，但换取更好的信号质量

### 2.4 相位调制（PM）

相位调制通过改变载波的瞬时相位来传输信息。

**调相信号的表达式**：

$$s(t) = A_c \cos[2\pi f_c t + k_p \cdot x(t)]$$

其中 $k_p$ 是相位灵敏度（rad/V）。

**PM与FM的关系**：

FM和PM本质上是同一种调制方式的不同表现形式：

- **FM**：频率偏移与调制信号成正比
- **PM**：相位偏移与调制信号成正比

两者的关系可以通过微分/积分相互转换：

$$\text{FM}: \phi(t) = 2\pi k_f \int_0^t x(\tau) d\tau$$

$$\text{PM}: \phi(t) = 2\pi k_p \cdot x(t)$$

如果对调制信号先积分再进行相位调制，得到的就是频率调制；反之，对调制信号先微分再进行相位调制，得到的就是FM。

**调制指数的对比**：

| 调制方式 | 调制指数公式 | 频偏关系 |
|----------|--------------|----------|
| FM | $\beta_{FM} = \frac{\Delta f}{f_m}$ | $\Delta f = k_f \cdot A_m$ |
| PM | $\beta_{PM} = k_p \cdot A_m$ | 相位偏移峰值 = $\beta_{PM}$ |

### 2.5 卡森法则（Carson's Rule）

卡森法则用于估算调频信号的近似带宽：

$$B_T = 2(\Delta f + f_m) = 2f_m(\beta + 1)$$

其中：
- $B_T$：调频信号的总带宽
- $\Delta f$：峰值频偏（最大频率偏移）
- $f_m$：调制信号的频率
- $\beta$：调制指数

**卡森法则的物理意义**：
- 调频信号的频谱由载波和无数边带组成
- 边带位于 $f_c \pm n f_m$ 处（n为整数）
- 幅度按贝塞尔函数J_n(β)变化
- 当 |n| > (β + 1) 时，J_n(β) 迅速衰减接近零

**不同β值下的FM带宽特性**：

| β值 | 调制类型 | 带宽近似 | 主要应用 |
|------|----------|----------|----------|
| β << 1 | 窄带FM | $B_T ≈ 2f_m$ | 话音通信 |
| β ≈ 1 | 中等带宽 | $B_T ≈ 4f_m$ | 通用通信 |
| β >> 1 | 宽带FM | $B_T ≈ 2\Delta f$ | FM广播（β=5） |

## 3. Python演示

下面我们使用Python来演示各种模拟调制方式的原理和特性。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

print("模拟调制演示环境已准备就绪")
```

### 3.1 幅度调制（AM）演示

```python
def demo_amplitude_modulation():
    """
    演示标准AM调制的原理和特性
    
    参数说明：
    - fc: 载波频率 (Hz)
    - fm: 调制信号频率 (Hz)
    - m: 调制度 (0 < m <= 1)
    - fs: 采样频率 (Hz)
    """
    
    # 参数设置
    fc = 1000      # 载波频率 1000Hz
    fm = 50        # 调制信号频率 50Hz
    m = 0.8        # 调制度 0.8（80%）
    fs = 10000     # 采样频率 10000Hz
    T = 0.1        # 信号持续时间 0.1秒
    
    t = np.linspace(0, T, int(fs * T))
    
    # 生成调制信号（单音正弦波）
    x_t = np.sin(2 * np.pi * fm * t)
    
    # 生成AM调制信号
    # s(t) = [1 + m * x(t)] * cos(2πfc t)
    am_signal = (1 + m * x_t) * np.cos(2 * np.pi * fc * t)
    
    # 计算频谱
    N = len(t)
    X_freq = np.fft.fft(x_t)
    AM_freq = np.fft.fft(am_signal)
    freq_axis = np.fft.fftfreq(N, 1/fs)
    
    # 只取正频率部分
    pos_mask = freq_axis > 0
    
    # 创建图形
    fig, axes = plt.subplots(3, 2, figsize=(14, 12))
    
    # 1. 调制信号时域
    axes[0, 0].plot(t * 1000, x_t, 'b-', linewidth=1.5)
    axes[0, 0].set_ylabel('幅度', fontsize=11)
    axes[0, 0].set_title('调制信号 x(t) = sin(2π×50t)', fontsize=12)
    axes[0, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].set_ylim(-1.5, 1.5)
    
    # 2. AM调制信号时域
    axes[0, 1].plot(t * 1000, am_signal, 'r-', linewidth=1.5)
    axes[0, 1].plot(t * 1000, 1 + m * x_t, 'g--', linewidth=1, alpha=0.7, label='包络 1+m·x(t)')
    axes[0, 1].plot(t * 1000, -(1 + m * x_t), 'g--', linewidth=1, alpha=0.7)
    axes[0, 1].set_ylabel('幅度', fontsize=11)
    axes[0, 1].set_title(f'AM调制信号 (m={m})', fontsize=12)
    axes[0, 1].set_xlabel('时间 (ms)', fontsize=11)
    axes[0, 1].legend(loc='upper right')
    axes[0, 1].grid(True, alpha=0.3)
    
    # 3. 调制信号频谱
    axes[1, 0].stem(freq_axis[pos_mask] / 1000, 
                   np.abs(X_freq[pos_mask]) / N * 2, 
                   markerfmt=' ', basefmt=' ')
    axes[1, 0].set_ylabel('幅度', fontsize=11)
    axes[1, 0].set_title('调制信号频谱', fontsize=12)
    axes[1, 0].set_xlabel('频率 (kHz)', fontsize=11)
    axes[1, 0].set_xlim(0, 0.3)
    axes[1, 0].grid(True, alpha=0.3)
    
    # 4. AM调制信号频谱
    axes[1, 1].stem(freq_axis[pos_mask] / 1000, 
                   np.abs(AM_freq[pos_mask]) / N * 2, 
                   markerfmt=' ', basefmt=' ')
    axes[1, 1].axvline(x=fc/1000, color='r', linestyle='--', alpha=0.7, label=f'fc={fc}Hz')
    axes[1, 1].axvline(x=(fc-fm)/1000, color='g', linestyle='--', alpha=0.7, label=f'fc-fm')
    axes[1, 1].axvline(x=(fc+fm)/1000, color='g', linestyle='--', alpha=0.7, label=f'fc+fm')
    axes[1, 1].set_ylabel('幅度', fontsize=11)
    axes[1, 1].set_title('AM调制信号频谱', fontsize=12)
    axes[1, 1].set_xlabel('频率 (kHz)', fontsize=11)
    axes[1, 1].set_xlim(0.5, 1.5)
    axes[1, 1].legend(loc='upper right')
    axes[1, 1].grid(True, alpha=0.3)
    
    # 5. 不同调制度的对比
    m_values = [0.3, 0.6, 1.0]
    colors = ['blue', 'green', 'red']
    for m_val, color in zip(m_values, colors):
        am_val = (1 + m_val * x_t) * np.cos(2 * np.pi * fc * t)
        axes[2, 0].plot(t * 1000, am_val, color=color, linewidth=1.5, 
                       label=f'm={m_val}', alpha=0.8)
    axes[2, 0].set_ylabel('幅度', fontsize=11)
    axes[2, 0].set_title('不同调制度的AM信号对比', fontsize=12)
    axes[2, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[2, 0].legend(loc='upper right')
    axes[2, 0].grid(True, alpha=0.3)
    
    # 6. 过调制演示
    m_over = 1.5  # 过调制度
    am_over = (1 + m_over * x_t) * np.cos(2 * np.pi * fc * t)
    axes[2, 1].plot(t * 1000, am_over, 'purple', linewidth=1.5)
    axes[2, 1].plot(t * 1000, 1 + m_over * x_t, 'r--', linewidth=1, alpha=0.7, label='包络')
    axes[2, 1].plot(t * 1000, -(1 + m_over * x_t), 'r--', linewidth=1, alpha=0.7)
    axes[2, 1].axhline(y=0, color='k', linewidth=0.5)
    axes[2, 1].set_ylabel('幅度', fontsize=11)
    axes[2, 1].set_title(f'过调制信号 (m={m_over} > 1) - 信号失真', fontsize=12)
    axes[2, 1].set_xlabel('时间 (ms)', fontsize=11)
    axes[2, 1].legend(loc='upper right')
    axes[2, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 打印AM调制特性总结
    print("\n幅度调制(AM)特性分析：")
    print("=" * 60)
    print(f"载波频率: {fc} Hz")
    print(f"调制信号频率: {fm} Hz")
    print(f"调制度: {m} ({m*100}%)")
    print("-" * 60)
    print("AM信号频谱组成：")
    print(f"  - 载波分量: {fc} Hz")
    print(f"  - 上边带: {fc + fm} Hz (USB)")
    print(f"  - 下边带: {fc - fm} Hz (LSB)")
    print("-" * 60)
    print("功率分配（单音调制）：")
    print(f"  - 载波功率占比: {1/(1+m**2/2)*100:.1f}%")
    print(f"  - 边带功率占比: {m**2/(2+m**2)*100:.1f}%")
    print("-" * 60)
    print("结论：")
    print("  - AM调制包含载波，功率效率较低")
    print("  - 调制度m越大，边带功率占比越高")
    print("  - 过调制(m>1)会导致信号失真，解调困难")
    print("=" * 60)
    
    return

demo_amplitude_modulation()
```

### 3.2 DSB-SC 和 SSB 调制演示

```python
def demo_dsb_ssbmodulation():
    """
    演示DSB-SC和SSB调制方式的对比
    
    参数说明：
    - fc: 载波频率
    - fm: 调制信号频率
    - fs: 采样频率
    """
    
    # 参数设置
    fc = 1000      # 载波频率 1000Hz
    fm = 100       # 调制信号频率 100Hz
    fs = 10000     # 采样频率 10000Hz
    T = 0.08       # 信号持续时间 0.08秒
    
    t = np.linspace(0, T, int(fs * T))
    
    # 生成调制信号
    x_t = np.sin(2 * np.pi * fm * t)
    
    # 1. 标准AM
    m = 0.8
    am_signal = (1 + m * x_t) * np.cos(2 * np.pi * fc * t)
    
    # 2. DSB-SC（抑制载波的双边带）
    dsb_signal = x_t * np.cos(2 * np.pi * fc * t)
    
    # 3. SSB（上边带）- 使用希尔伯特变换近似
    # s_SSB(t) = x(t)*cos(wc*t) - x_hat(t)*sin(wc*t)
    from scipy.signal import hilbert
    x_hilbert = np.imag(hilbert(x_t))  # 希尔伯特变换（90度相移）
    ssb_signal = x_t * np.cos(2 * np.pi * fc * t) - x_hilbert * np.sin(2 * np.pi * fc * t)
    
    # 计算各信号的频谱
    N = len(t)
    freq_axis = np.fft.fftfreq(N, 1/fs)
    pos_mask = freq_axis > 0
    
    AM_freq = np.fft.fft(am_signal)
    DSB_freq = np.fft.fft(dsb_signal)
    SSB_freq = np.fft.fft(ssb_signal)
    
    # 创建图形
    fig, axes = plt.subplots(3, 2, figsize=(14, 12))
    
    # AM时域
    axes[0, 0].plot(t * 1000, am_signal, 'b-', linewidth=1.5)
    axes[0, 0].set_ylabel('幅度', fontsize=11)
    axes[0, 0].set_title('标准AM信号 (m=0.8)', fontsize=12)
    axes[0, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[0, 0].grid(True, alpha=0.3)
    
    # AM频谱
    axes[0, 1].stem(freq_axis[pos_mask] / 1000, 
                   np.abs(AM_freq[pos_mask]) / N * 2, 
                   markerfmt=' ', basefmt=' ')
    axes[0, 1].set_ylabel('幅度', fontsize=11)
    axes[0, 1].set_title('AM信号频谱（含载波）', fontsize=12)
    axes[0, 1].set_xlabel('频率 (kHz)', fontsize=11)
    axes[0, 1].set_xlim(0.5, 1.5)
    axes[0, 1].grid(True, alpha=0.3)
    
    # DSB时域
    axes[1, 0].plot(t * 1000, dsb_signal, 'g-', linewidth=1.5)
    axes[1, 0].set_ylabel('幅度', fontsize=11)
    axes[1, 0].set_title('DSB-SC信号（抑制载波双边带）', fontsize=12)
    axes[1, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[1, 0].grid(True, alpha=0.3)
    
    # DSB频谱
    axes[1, 1].stem(freq_axis[pos_mask] / 1000, 
                   np.abs(DSB_freq[pos_mask]) / N * 2, 
                   markerfmt=' ', basefmt=' ')
    axes[1, 1].set_ylabel('幅度', fontsize=11)
    axes[1, 1].set_title('DSB-SC信号频谱（无载波）', fontsize=12)
    axes[1, 1].set_xlabel('频率 (kHz)', fontsize=11)
    axes[1, 1].set_xlim(0.5, 1.5)
    axes[1, 1].grid(True, alpha=0.3)
    
    # SSB时域
    axes[2, 0].plot(t * 1000, ssb_signal, 'r-', linewidth=1.5)
    axes[2, 0].set_ylabel('幅度', fontsize=11)
    axes[2, 0].set_title('SSB信号（单边带）', fontsize=12)
    axes[2, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[2, 0].grid(True, alpha=0.3)
    
    # SSB频谱
    axes[2, 1].stem(freq_axis[pos_mask] / 1000, 
                   np.abs(SSB_freq[pos_mask]) / N * 2, 
                   markerfmt=' ', basefmt=' ')
    axes[2, 1].set_ylabel('幅度', fontsize=11)
    axes[2, 1].set_title('SSB信号频谱（仅上边带）', fontsize=12)
    axes[2, 1].set_xlabel('频率 (kHz)', fontsize=11)
    axes[2, 1].set_xlim(0.5, 1.5)
    axes[2, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 打印对比总结
    print("\n幅度调制方式对比：")
    print("=" * 70)
    print(f"{'调制方式':<15} | {'载波':<10} | {'带宽':<10} | {'功率效率':<15} | {'解调难度'}")
    print("-" * 70)
    print(f"{'标准AM':<15} | {'有':<10} | {'2fm':<10} | {'低':<15} | {'简单（包络检波）'}")
    print(f"{'DSB-SC':<15} | {'无':<10} | {'2fm':<10} | {'高':<15} | {'需要同步载波'}")
    print(f"{'SSB':<15} | {'无':<10} | {'fm':<10} | {'最高':<15} | {'需要同步载波'}")
    print("-" * 70)
    print("\n应用场景：")
    print("  - 标准AM：广播Radio（简单接收机）")
    print("  - DSB-SC：部分通信系统")
    print("  - SSB：短波通信、军事通信（节省带宽）")
    print("=" * 70)
    
    return

demo_dsb_ssbmodulation()
```

### 3.3 频率调制（FM）演示

```python
def demo_frequency_modulation():
    """
    演示频率调制(FM)的原理和特性
    
    参数说明：
    - fc: 载波频率
    - fm: 调制信号频率
    - kf: 频率灵敏度
    - beta: 调制指数
    """
    
    # 参数设置
    fc = 2000      # 载波频率 2000Hz
    fm = 200       # 调制信号频率 200Hz
    kf = 500       # 频率灵敏度 500Hz/V
    fs = 50000     # 采样频率 50000Hz
    T = 0.05       # 信号持续时间 0.05秒
    
    t = np.linspace(0, T, int(fs * T))
    
    # 生成调制信号
    x_t = np.sin(2 * np.pi * fm * t)
    
    # 计算调制指数
    beta = kf / fm  # Δf / fm
    
    # 计算峰值频偏
    delta_f = kf * 1.0  # 最大频偏（调制幅度为1）
    
    # 生成FM信号
    # s(t) = Acos(2πfc t + 2πkf∫x(t)dt)
    instantaneous_phase = 2 * np.pi * fc * t + 2 * np.pi * kf * np.cumsum(x_t) / fs
    fm_signal = np.cos(instantaneous_phase)
    
    # 计算瞬时频率（用于验证）
    inst_freq = fc + kf * x_t
    
    # 计算频谱
    N = len(t)
    FM_freq = np.fft.fft(fm_signal)
    freq_axis = np.fft.fftfreq(N, 1/fs)
    pos_mask = freq_axis > 0
    
    # 创建图形
    fig, axes = plt.subplots(3, 2, figsize=(14, 12))
    
    # 1. 调制信号
    axes[0, 0].plot(t * 1000, x_t, 'b-', linewidth=1.5)
    axes[0, 0].set_ylabel('幅度', fontsize=11)
    axes[0, 0].set_title('调制信号 x(t)', fontsize=12)
    axes[0, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].set_ylim(-1.5, 1.5)
    
    # 2. FM信号时域
    axes[0, 1].plot(t * 1000, fm_signal, 'r-', linewidth=1.5)
    axes[0, 1].set_ylabel('幅度', fontsize=11)
    axes[0, 1].set_title(f'FM调制信号 (β={beta:.1f})', fontsize=12)
    axes[0, 1].set_xlabel('时间 (ms)', fontsize=11)
    axes[0, 1].grid(True, alpha=0.3)
    
    # 3. 瞬时频率变化
    axes[1, 0].plot(t * 1000, inst_freq / 1000, 'g-', linewidth=1.5)
    axes[1, 0].axhline(y=fc/1000, color='r', linestyle='--', alpha=0.7, label=f'fc={fc}Hz')
    axes[1, 0].axhline(y=(fc+delta_f)/1000, color='orange', linestyle='--', alpha=0.5)
    axes[1, 0].axhline(y=(fc-delta_f)/1000, color='orange', linestyle='--', alpha=0.5)
    axes[1, 0].set_ylabel('频率 (kHz)', fontsize=11)
    axes[1, 0].set_title('瞬时频率变化', fontsize=12)
    axes[1, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[1, 0].legend(loc='upper right')
    axes[1, 0].grid(True, alpha=0.3)
    
    # 4. FM频谱
    axes[1, 1].stem(freq_axis[pos_mask] / 1000, 
                   np.abs(FM_freq[pos_mask]) / N * 2, 
                   markerfmt=' ', basefmt=' ')
    axes[1, 1].set_ylabel('幅度', fontsize=11)
    axes[1, 1].set_title('FM调制信号频谱', fontsize=12)
    axes[1, 1].set_xlabel('频率 (kHz)', fontsize=11)
    axes[1, 1].set_xlim(1, 3)
    axes[1, 1].grid(True, alpha=0.3)
    
    # 5. 不同调制指数的对比
    beta_values = [0.5, 2.0, 5.0]
    colors = ['blue', 'green', 'red']
    
    for beta_val, color in zip(beta_values, colors):
        delta_f_val = beta_val * fm
        inst_phase = 2 * np.pi * fc * t + 2 * np.pi * delta_f_val/fm * np.sin(2*np.pi*fm*t)
        fm_temp = np.cos(inst_phase)
        FM_temp = np.fft.fft(fm_temp)
        
        axes[2, 0].plot(t * 1000, fm_temp, color=color, linewidth=1, 
                       label=f'β={beta_val}', alpha=0.8)
    
    axes[2, 0].set_ylabel('幅度', fontsize=11)
    axes[2, 0].set_title('不同调制指数β的FM信号', fontsize=12)
    axes[2, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[2, 0].legend(loc='upper right')
    axes[2, 0].grid(True, alpha=0.3)
    
    # 6. 不同β的频谱对比
    for beta_val, color in zip(beta_values, colors):
        delta_f_val = beta_val * fm
        inst_phase = 2 * np.pi * fc * t + 2 * np.pi * delta_f_val/fm * np.sin(2*np.pi*fm*t)
        fm_temp = np.cos(inst_phase)
        FM_temp = np.fft.fft(fm_temp)
        
        # 绘制正频率部分（归一化）
        freq_mask = (freq_axis > 0) & (freq_axis < 4000)
        axes[2, 1].plot(freq_axis[freq_mask] / 1000, 
                       np.abs(FM_temp[freq_mask]) / N * 2, 
                       color=color, linewidth=1.5, label=f'β={beta_val}')
    
    axes[2, 1].set_ylabel('幅度', fontsize=11)
    axes[2, 1].set_title('不同调制指数β的FM频谱', fontsize=12)
    axes[2, 1].set_xlabel('频率 (kHz)', fontsize=11)
    axes[2, 1].set_xlim(1.5, 2.5)
    axes[2, 1].legend(loc='upper right')
    axes[2, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 计算卡森带宽
    carsons_bw = 2 * (delta_f + fm)
    
    print("\n频率调制(FM)特性分析：")
    print("=" * 60)
    print(f"载波频率 fc: {fc} Hz")
    print(f"调制信号频率 fm: {fm} Hz")
    print(f"频率灵敏度 kf: {kf} Hz/V")
    print(f"峰值频偏 Δf: {delta_f} Hz")
    print(f"调制指数 β: {beta:.2f}")
    print("-" * 60)
    print(f"卡森法则带宽: B_T = 2(Δf + fm) = {carsons_bw} Hz")
    print(f"理论带宽（贝塞尔展开）: ≈ 2(β + 1)·fm = {2*(beta+1)*fm} Hz")
    print("-" * 60)
    print("FM调制特性：")
    print("  - 幅度恒定，抗幅度噪声能力强")
    print("  - β越大，频谱扩展越宽，边带越多")
    print("  - FM广播使用β≈5，话音通信使用β≈2")
    print("  - FM带宽效率低于AM，但抗干扰能力更强")
    print("=" * 60)
    
    return

demo_frequency_modulation()
```

### 3.4 FM解调演示

```python
def demo_fm_demodulation():
    """
    演示FM信号的解调方法
    使用微分器法（鉴频器）进行解调
    
    解调原理：
    - FM: s(t) = Acos(2πfc t + φ(t))
    - 微分后: ds/dt ∝ sin(2πfc t + φ(t))·(2πfc + dφ/dt)
    - 包络检波提取 dφ/dt 即为原调制信号
    """
    
    # 参数设置
    fc = 2000      # 载波频率
    fm = 200        # 调制信号频率
    kf = 500        # 频率灵敏度
    fs = 50000      # 采样频率
    T = 0.08        # 信号持续时间
    
    t = np.linspace(0, T, int(fs * T))
    
    # 生成调制信号
    x_t = np.sin(2 * np.pi * fm * t)
    
    # 生成FM信号
    beta = kf / fm
    instantaneous_phase = 2 * np.pi * fc * t + 2 * np.pi * kf * np.cumsum(x_t) / fs
    fm_signal = np.cos(instantaneous_phase)
    
    # 添加一些噪声模拟真实信道
    np.random.seed(42)
    noise = 0.05 * np.random.randn(len(t))
    fm_noisy = fm_signal + noise
    
    # FM解调方法1：微分 + 包络检波
    # 微分
    fm_derivative = np.diff(fm_noisy) * fs  # 近似微分
    # 包络检波（取绝对值后低通滤波）
    envelope = np.abs(fm_derivative)
    # 去除直流分量
    from scipy.signal import butter, filtfilt
    b, a = butter(4, fm / (fs/2), btype='low')
    x_demod = filtfilt(b, a, envelope)
    # 归一化
    x_demod = x_demod / np.max(np.abs(x_demod))
    
    # 简化的微分-鉴频方法
    analytic_signal = signal.hilbert(fm_noisy[:-1])
    inst_phase = np.angle(analytic_signal)
    inst_phase_unwrapped = np.unwrap(inst_phase)
    # 差分得到瞬时频率
    inst_freq = np.diff(inst_phase_unwrapped) * fs / (2*np.pi)
    # 去除直流和低频
    x_demod2 = inst_freq - np.mean(inst_freq)
    x_demod2 = x_demod2 / np.max(np.abs(x_demod2))
    
    # 创建图形
    fig, axes = plt.subplots(4, 1, figsize=(14, 14))
    
    # 1. 原始调制信号
    axes[0].plot(t * 1000, x_t, 'b-', linewidth=1.5)
    axes[0].set_ylabel('幅度', fontsize=11)
    axes[0].set_title('原始调制信号', fontsize=12)
    axes[0].grid(True, alpha=0.3)
    
    # 2. FM已调信号（含噪声）
    axes[1].plot(t * 1000, fm_noisy, 'r-', linewidth=1)
    axes[1].set_ylabel('幅度', fontsize=11)
    axes[1].set_title('FM已调信号（加噪声）', fontsize=12)
    axes[1].grid(True, alpha=0.3)
    
    # 3. 解调信号（方法1）
    t_short = t[:-1]
    axes[2].plot(t_short * 1000, x_demod[:len(t_short)], 'g-', linewidth=1.5, 
                 label='微分+包络法')
    axes[2].set_ylabel('幅度', fontsize=11)
    axes[2].set_title('FM解调信号（微分+包络检波法）', fontsize=12)
    axes[2].legend(loc='upper right')
    axes[2].grid(True, alpha=0.3)
    
    # 4. 解调信号（方法2 - 相位法）
    t_phase = t[:len(x_demod2)]
    axes[3].plot(t_phase * 1000, x_demod2, 'm-', linewidth=1.5, label='相位法')
    axes[3].plot(t * 1000, x_t, 'b--', linewidth=1, alpha=0.7, label='原始信号')
    axes[3].set_ylabel('幅度', fontsize=11)
    axes[3].set_xlabel('时间 (ms)', fontsize=11)
    axes[3].set_title('FM解调信号（相位法/鉴频器）', fontsize=12)
    axes[3].legend(loc='upper right')
    axes[3].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 计算解调误差
    # 使用相位法结果计算SNR改善
    signal_power = np.var(x_t)
    noise_power_1 = np.var(x_demod[:len(t)] - x_t)
    noise_power_2 = np.var(x_demod2 - x_t[:len(x_demod2)])
    
    print("\nFM解调分析：")
    print("=" * 60)
    print("常用FM解调方法：")
    print("  1. 微分器法：ds/dt → 包络检波")
    print("  2. 相位鉴频器：提取瞬时相位/频率")
    print("  3. 锁相环(PLL)解调：跟踪载波相位（性能最佳）")
    print("  4. 科斯塔斯环(Costas Loop)：载波同步")
    print("-" * 60)
    print("FM解调的特点：")
    print("  - FM为恒包络调制，解调不依赖幅度")
    print("  - 噪声主要影响幅度，对频率影响较小")
    print("  - 具有'噪声抑制效应'，窄带FM优于宽带FM")
    print("  - 但宽带FM的信号质量更好（更好的信噪比）")
    print("=" * 60)
    
    return

demo_fm_demodulation()
```

### 3.5 PM调制演示及与FM对比

```python
def demo_pm_modulation():
    """
    演示相位调制(PM)及其与FM的关系
    
    PM: s(t) = Acos(2πfc t + kp·x(t))
    FM: s(t) = Acos(2πfc t + 2πkf∫x(t)dt)
    
    两者关系：对调制信号先积分后PM = FM
              对调制信号先微分后FM = PM
    """
    
    # 参数设置
    fc = 1000      # 载波频率
    fm = 100       # 调制信号频率
    kp = 2.0       # 相位灵敏度 (rad/V)
    fs = 20000     # 采样频率
    T = 0.1        # 信号持续时间
    
    t = np.linspace(0, T, int(fs * T))
    
    # 生成调制信号（复杂信号：两个频率成分）
    x_t = 0.5 * np.sin(2 * np.pi * fm * t) + 0.5 * np.sin(2 * np.pi * 2*fm * t)
    
    # PM调制
    pm_signal = np.cos(2 * np.pi * fc * t + kp * x_t)
    
    # FM调制（等效参数）
    kf = 500  # 选择使调制指数相近的值
    fm_signal = np.cos(2 * np.pi * fc * t + 2 * np.pi * kf * np.cumsum(x_t) / fs)
    
    # 计算瞬时相位和频率
    pm_phase = 2 * np.pi * fc * t + kp * x_t
    pm_inst_freq = fc + kp/(2*np.pi) * np.diff(x_t) * fs
    
    fm_phase = 2 * np.pi * fc * t + 2 * np.pi * kf * np.cumsum(x_t) / fs
    fm_inst_freq = fc + kf * x_t
    
    # 计算频谱
    N = len(t)
    freq_axis = np.fft.fftfreq(N, 1/fs)
    pos_mask = freq_axis > 0
    
    PM_freq = np.fft.fft(pm_signal)
    FM_freq = np.fft.fft(fm_signal)
    
    # 创建图形
    fig, axes = plt.subplots(3, 2, figsize=(14, 12))
    
    # 1. 调制信号
    axes[0, 0].plot(t * 1000, x_t, 'b-', linewidth=1.5)
    axes[0, 0].set_ylabel('幅度', fontsize=11)
    axes[0, 0].set_title('调制信号 x(t)', fontsize=12)
    axes[0, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[0, 0].grid(True, alpha=0.3)
    
    # 2. PM信号时域
    axes[0, 1].plot(t * 1000, pm_signal, 'r-', linewidth=1.5)
    axes[0, 1].set_ylabel('幅度', fontsize=11)
    axes[0, 1].set_title(f'PM调制信号 (kp={kp} rad/V)', fontsize=12)
    axes[0, 1].set_xlabel('时间 (ms)', fontsize=11)
    axes[0, 1].grid(True, alpha=0.3)
    
    # 3. FM信号时域
    axes[1, 0].plot(t * 1000, fm_signal, 'g-', linewidth=1.5)
    axes[1, 0].set_ylabel('幅度', fontsize=11)
    axes[1, 0].set_title(f'FM调制信号 (kf={kf} Hz/V)', fontsize=12)
    axes[1, 0].set_xlabel('时间 (ms)', fontsize=11)
    axes[1, 0].grid(True, alpha=0.3)
    
    # 4. PM频谱
    axes[1, 1].stem(freq_axis[pos_mask] / 1000, 
                   np.abs(PM_freq[pos_mask]) / N * 2, 
                   markerfmt=' ', basefmt=' ')
    axes[1, 1].set_ylabel('幅度', fontsize=11)
    axes[1, 1].set_title('PM信号频谱', fontsize=12)
    axes[1, 1].set_xlabel('频率 (kHz)', fontsize=11)
    axes[1, 1].set_xlim(0.5, 1.5)
    axes[1, 1].grid(True, alpha=0.3)
    
    # 5. FM频谱
    axes[2, 0].stem(freq_axis[pos_mask] / 1000, 
                   np.abs(FM_freq[pos_mask]) / N * 2, 
                   markerfmt=' ', basefmt=' ')
    axes[2, 0].set_ylabel('幅度', fontsize=11)
    axes[2, 0].set_title('FM信号频谱', fontsize=12)
    axes[2, 0].set_xlabel('频率 (kHz)', fontsize=11)
    axes[2, 0].set_xlim(0.5, 1.5)
    axes[2, 0].grid(True, alpha=0.3)
    
    # 6. PM与FM的瞬时频率对比
    t_inst = t[:-1]
    axes[2, 1].plot(t_inst * 1000, pm_inst_freq / 1000, 'r-', linewidth=1.5, 
                   label='PM瞬时频率', alpha=0.8)
    axes[2, 1].plot(t * 1000, fm_inst_freq / 1000, 'g--', linewidth=1.5, 
                   label='FM瞬时频率', alpha=0.8)
    axes[2, 1].axhline(y=fc/1000, color='b', linestyle=':', alpha=0.5)
    axes[2, 1].set_ylabel('频率 (kHz)', fontsize=11)
    axes[2, 1].set_xlabel('时间 (ms)', fontsize=11)
    axes[2, 1].set_title('PM与FM的瞬时频率对比', fontsize=12)
    axes[2, 1].legend(loc='upper right')
    axes[2, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n相位调制(PM)与频率调制(FM)对比：")
    print("=" * 60)
    print("数学表达：")
    print("  PM: s(t) = A·cos(2πfct + kp·x(t))")
    print("  FM: s(t) = A·cos(2πfct + 2πkf∫x(t)dt)")
    print("-" * 60)
    print("关系：")
    print("  - FM是对调制信号积分后的PM")
    print("  - PM是对调制信号微分后的FM")
    print("  - 本质上，两者都是角度调制")
    print("-" * 60)
    print("瞬时频率特性：")
    print("  - PM：瞬时频率与调制信号的微分成正比")
    print("  - FM：瞬时频率与调制信号成正比")
    print("-" * 60)
    print("应用：")
    print("  - FM更常用（话音通信、广播）")
    print("  - PM用于某些数字调制（如MPSK）的基础")
    print("=" * 60)
    
    return

demo_pm_modulation()
```

### 3.6 卡森法则演示

```python
def demo_carsons_rule():
    """
    演示卡森法则及其在不同调制指数下的适用性
    
    卡森法则: BT = 2(Δf + fm) = 2fm(β + 1)
    
    当β较大时，卡森法则给出约98%的功率覆盖
    """
    
    # 参数设置
    fc = 5000      # 载波频率 5kHz
    fm = 500       # 调制信号频率 500Hz
    fs = 100000    # 采样频率
    T = 0.1        # 信号持续时间
    
    t = np.linspace(0, T, int(fs * T))
    
    # 调制信号
    x_t = np.sin(2 * np.pi * fm * t)
    
    # 测试不同的调制指数
    beta_values = [0.5, 1.0, 2.0, 5.0]
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    axes = axes.flatten()
    
    for idx, beta in enumerate(beta_values):
        # 计算FM信号
        delta_f = beta * fm
        inst_phase = 2 * np.pi * fc * t + beta * np.sin(2*np.pi*fm*t)
        fm_signal = np.cos(inst_phase)
        
        # 计算频谱
        N = len(t)
        FM_freq = np.fft.fft(fm_signal)
        freq_axis = np.fft.fftfreq(N, 1/fs)
        pos_mask = freq_axis > 0
        
        # 归一化频谱幅度
        spectrum = np.abs(FM_freq[pos_mask]) / N * 2
        total_power = np.sum(spectrum**2)
        
        # 卡森法则带宽
        BT = 2 * (delta_f + fm)
        
        # 计算卡森带宽内的功率占比
        freq_pos = freq_axis[pos_mask]
        inside_mask = (freq_pos > fc - BT/2) & (freq_pos < fc + BT/2)
        inside_power = np.sum(spectrum[inside_mask]**2)
        power_ratio = inside_power / total_power * 100
        
        # 绘制频谱
        axes[idx].stem(freq_pos / 1000, spectrum, markerfmt=' ', basefmt=' ')
        axes[idx].axvline(x=(fc-BT/2)/1000, color='r', linestyle='--', 
                         linewidth=2, label=f'fc±BT/2')
        axes[idx].axvline(x=(fc+BT/2)/1000, color='r', linestyle='--', linewidth=2)
        
        # 标注实际带宽边界（功率99%以上）
        cumsum_power = np.cumsum(spectrum**2)
        cumsum_power = cumsum_power / cumsum_power[-1]
        low_idx = np.searchsorted(cumsum_power, 0.005)  # 0.5%下限
        high_idx = np.searchsorted(cumsum_power, 0.995)  # 99.5%上限
        actual_bw = freq_pos[high_idx] - freq_pos[low_idx]
        
        axes[idx].axvline(x=freq_pos[low_idx]/1000, color='g', linestyle=':', 
                         linewidth=1.5, alpha=0.7)
        axes[idx].axvline(x=freq_pos[high_idx]/1000, color='g', linestyle=':', 
                         linewidth=1.5, alpha=0.7)
        
        axes[idx].set_ylabel('幅度', fontsize=11)
        axes[idx].set_xlabel('频率 (kHz)', fontsize=11)
        axes[idx].set_title(f'β={beta}, Δf={delta_f}Hz, BT={BT}Hz\n'
                           f'卡森带宽内功率: {power_ratio:.1f}%', fontsize=11)
        axes[idx].set_xlim(3, 7)
        axes[idx].legend(loc='upper right', fontsize=9)
        axes[idx].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 打印卡森法则分析
    print("\n卡森法则(Carson's Rule)分析：")
    print("=" * 70)
    print("公式: BT = 2(Δf + fm) = 2fm(β + 1)")
    print("-" * 70)
    print("\n不同调制指数下的带宽覆盖：")
    print(f"{'β':<8} | {'Δf (Hz)':<12} | {'卡森带宽':<12} | {'功率覆盖':<10}")
    print("-" * 70)
    
    for beta in [0.1, 0.5, 1.0, 2.0, 5.0, 10.0]:
        delta_f = beta * fm
        BT = 2 * (delta_f + fm)
        # 近似功率覆盖（理论值约为 1 - 2/β for large β）
        if beta < 0.5:
            power_cover = 85 + beta * 20
        elif beta < 2:
            power_cover = 90 + (beta - 0.5) * 4
        else:
            power_cover = min(98 + 1.5/beta, 99.9)
        print(f"{beta:<8.1f} | {delta_f:<12.1f} | {BT:<12.1f} | {power_cover:.1f}%")
    
    print("-" * 70)
    print("\n卡森法则说明：")
    print("  - β << 1（窄带FM）：BT ≈ 2fm，边带较少")
    print("  - β ≈ 1（过渡区）：BT ≈ 4fm，边带增多")
    print("  - β >> 1（宽带FM）：BT ≈ 2Δf，主导项变化")
    print("  - 实际工程中，常用卡森带宽作为FM信号带宽的近似")
    print("=" * 70)
    
    return

demo_carsons_rule()
```

### 4.1 调制方式的对比总结

```python
def modulation_summary():
    """
    总结各种模拟调制方式的特点和应用
    """
    
    print("=" * 80)
    print("模拟调制方式综合对比")
    print("=" * 80)
    
    print("\n【1】幅度调制家族")
    print("-" * 60)
    print("\n  标准AM (DSB-FC):")
    print("    表达式: s(t) = [1 + m·x(t)]·cos(2πfct)")
    print("    优点: 接收机简单（包络检波）")
    print("    缺点: 功率效率低（载波占大部分功率）")
    print("    应用: AM广播")
    
    print("\n  DSB-SC (抑制载波双边带):")
    print("    表达式: s(t) = x(t)·cos(2πfct)")
    print("    优点: 功率效率高（无载波）")
    print("    缺点: 解调需要同步载波")
    print("    应用: 电视图像信号、某些通信系统")
    
    print("\n  SSB (单边带):")
    print("    表达式: s(t) = x(t)·cos(2πfct) - x_hat(t)·sin(2πfct)")
    print("    优点: 带宽利用率最高（仅为基带带宽）")
    print("    缺点: 解调需要同步载波，技术难度高")
    print("    应用: 短波通信、军事通信、电话系统")
    
    print("\n【2】角度调制家族")
    print("-" * 60)
    
    print("\n  FM (频率调制):")
    print("    表达式: s(t) = Acos(2πfct + 2πkf∫x(t)dt)")
    print("    优点: 抗干扰能力强，恒包络（放大效率高）")
    print("    缺点: 带宽利用率低，边带复杂")
    print("    应用: FM广播、话音通信、卫星通信")
    
    print("\n  PM (相位调制):")
    print("    表达式: s(t) = Acos(2πfct + kp·x(t))")
    print("    优点: 与FM类似，可通过积分相互转换")
    print("    缺点: 对相位噪声敏感")
    print("    应用: 数字调制基础（MPSK、QPSK等）")
    
    print("\n【3】性能指标对比")
    print("-" * 60)
    print("\n  调制方式  | 带宽效率  | 功率效率  | 抗噪性能 | 接收复杂度")
    print("  " + "-" * 56)
    print("  标准AM   |   低     |   低     |   差     |   简单")
    print("  DSB-SC   |   中     |   高     |   中     |   中等")
    print("  SSB      |   高     |   高     |   中     |   复杂")
    print("  FM(β=5)  |   低     |   中     |   优     |   中等")
    print("  PM       |   低     |   中     |   优     |   中等")
    
    print("\n【4】调制方式选择原则")
    print("-" * 60)
    print("  - 带宽受限：选择SSB（最高频谱效率）")
    print("  - 功率受限：选择DSB-SC或SSB（无载波功率浪费）")
    print("  - 噪声环境：选择FM（抗噪声能力强）")
    print("  - 接收机简单：选择标准AM（包络检波）")
    print("  - 功放效率：选择恒包络调制（FM/PM）")
    
    print("\n" + "=" * 80)
    return

modulation_summary()
```

### 4.2 调制指数与带宽的关系

```python
def bandwidth_vs_modulation_index():
    """
    分析调制指数与传输带宽的关系
    展示不同调制指数下FM信号的带宽扩展特性
    """
    
    # 参数设置
    fm = 1.0        # 归一化调制频率
    beta_range = np.linspace(0.1, 10, 100)
    
    # 卡森法则带宽
    BT_carson = 2 * fm * (beta_range + 1)
    
    # 实际带宽（近似：包含99%功率的带宽）
    # 对于FM，99%功率带宽近似为 2*(β+1.5)*fm
    BT_99 = 2 * fm * (beta_range + 1.5)
    
    # 创建图形
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # 1. 带宽 vs 调制指数
    axes[0].plot(beta_range, BT_carson, 'b-', linewidth=2, label='卡森法则带宽 BT')
    axes[0].plot(beta_range, BT_99, 'r--', linewidth=2, label='99%功率带宽')
    axes[0].plot(beta_range, 2*fm*beta_range, 'g:', linewidth=2, label='2Δf (β>>1渐近线)')
    axes[0].plot(beta_range, 2*fm*np.ones_like(beta_range), 'k:', linewidth=1, alpha=0.5)
    
    # 标注区域
    axes[0].axvspan(0, 0.5, alpha=0.1, color='blue', label='窄带FM')
    axes[0].axvspan(0.5, 5, alpha=0.1, color='yellow')
    axes[0].axvspan(5, 10, alpha=0.1, color='red', label='宽带FM')
    
    axes[0].set_xlabel('调制指数 β', fontsize=11)
    axes[0].set_ylabel('归一化带宽 BT/(2fm)', fontsize=11)
    axes[0].set_title('FM带宽与调制指数的关系', fontsize=12)
    axes[0].legend(loc='upper left')
    axes[0].grid(True, alpha=0.3)
    axes[0].set_xlim(0, 10)
    axes[0].set_ylim(0, 12)
    
    # 2. 功率分布 vs 调制指数
    # 不同β下的边带功率分布
    beta_examples = [0.5, 1, 2, 5]
    colors = ['blue', 'green', 'orange', 'red']
    
    for beta, color in zip(beta_examples, colors):
        # 计算边带位置（相对于载波的次数）
        # 近似：95%功率在 |n| <= β+1 范围内
        n_range = np.arange(0, int(beta+2))
        # 近似功率分布（使用第一类贝塞尔函数平方的包络）
        # J0(β)^2 + 2*sum(Jn(β)^2 for n>=1) ≈ 1
        # 简化近似：主要边带贡献
        
        # 使用scipy计算贝塞尔函数
        from scipy.special import jn
        J = np.array([jn(n, beta) for n in range(int(beta+3))])
        power_ratio = np.cumsum(J**2) / np.sum(J**2)
        
        n_plot = np.arange(1, len(J))
        axes[1].plot(n_plot, power_ratio[1:] * 100, '-o', color=color, 
                     markersize=4, label=f'β={beta}', linewidth=1.5)
    
    axes[1].axhline(y=95, color='k', linestyle='--', alpha=0.5, label='95%功率线')
    axes[1].axhline(y=99, color='k', linestyle=':', alpha=0.5, label='99%功率线')
    axes[1].set_xlabel('边带次数 n', fontsize=11)
    axes[1].set_ylabel('累积功率百分比 (%)', fontsize=11)
    axes[1].set_title('FM边带功率累积分布', fontsize=12)
    axes[1].legend(loc='lower right')
    axes[1].grid(True, alpha=0.3)
    axes[1].set_xlim(0, 8)
    axes[1].set_ylim(0, 105)
    
    plt.tight_layout()
    plt.show()
    
    print("\n调制指数与带宽的关系：")
    print("=" * 60)
    print("\n【窄带FM (β < 0.5)】")
    print("  - 带宽: BT ≈ 2fm（与AM相近）")
    print("  - 特性: 频谱简单，主要边带数量少")
    print("  - 应用: 对讲机、某些低功率通信设备")
    
    print("\n【宽带FM (β > 1)】")
    print("  - 带宽: BT ≈ 2Δf = 2βfm（频偏主导）")
    print("  - 特性: 频谱扩展，多个边带，噪声性能好")
    print("  - 应用: FM广播(β≈5)、高保真音频传输")
    
    print("\n【设计权衡】")
    print("  - β↑ → 抗噪声性能↑ 但 带宽↑")
    print("  - β↓ → 带宽↓ 但 抗噪声性能↓")
    print("  - 工程中根据 信噪比需求 和 带宽限制 选择β")
    print("=" * 60)
    
    return

bandwidth_vs_modulation_index()
```

## 总结

本notebook介绍了模拟调制的基本概念和实现方法：

- **调制的意义**：解决无线传输、频谱效率、天线尺寸等问题
- **幅度调制(AM/DSB/SSB)**：通过改变载波幅度传输信息，带宽效率逐级提高
- **频率调制(FM)**：通过改变载波频率传输信息，抗噪声能力强
- **相位调制(PM)**：与FM本质相同，是角度调制的另一种形式
- **卡森法则**：BT = 2(Δf + fm) = 2fm(β + 1)，用于估算FM带宽
- **调制指数**：β是FM设计的关键参数，决定带宽与性能的权衡

模拟调制是现代通信系统的基础，虽然数字调制日益普及，但理解模拟调制的原理对于深入学习通信系统至关重要。
