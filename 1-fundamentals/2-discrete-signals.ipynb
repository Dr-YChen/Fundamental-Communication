## 2. 概念讲解

### 2.1 连续时间信号 vs 离散时间信号

**连续时间信号（Continuous-Time Signal）**
- 定义：在任意时间点都有定义的信号
- 表示：$x_c(t)$，其中$t \in \mathbb{R}$
- 例子：自然界的声音、温度变化等

**离散时间信号（Discrete-Time Signal）**
- 定义：仅在特定时间点有定义的信号
- 表示：$x[n]$，其中$n \in \mathbb{Z}$（整数索引）
- 例子：数字采样后的信号、数字图像的像素值

**关键区别**
| 特性 | 连续信号 | 离散信号 |
|------|----------|----------|
| 时间轴 | 连续（任意实数） | 离散（整数索引） |
| 幅度 | 连续或离散 | 通常离散 |
| 处理方式 | 模拟电路 | 数字处理 |
| 存储 | 连续介质 | 有限精度 |

### 2.2 采样过程

采样是将连续时间信号转换为离散时间信号的过程：

$$x[n] = x_c(nT) = x_c(t) \big|_{t=nT}$$

其中：
- $T$：采样周期（秒），两个采样点之间的时间间隔
- $f_s = 1/T$：采样频率（Hz），每秒采集的样本数
- $n$：整数索引，表示第n个采样点

**采样过程示意图**
```
连续信号 x_c(t)  -->  [采样器]  -->  离散信号 x[n]
                              T
```

### 2.3 Nyquist采样定理

**Nyquist采样定理**是数字信号处理中最重要的定理之一：

> 对于一个带宽限制在$0 \sim f_{max}$Hz的连续信号，如果采样频率$f_s$满足：
>
> $$f_s \geq 2 f_{max}$$
>
> 那么可以无失真地重建原始连续信号。

**关键参数**
- $f_{max}$：信号的最高频率成分
- $f_s/2$：Nyquist频率（也叫折叠频率）
- $f_{max}$必须小于$f_s/2$才能避免混叠

| 采样情况 | 条件 | 结果 |
|----------|------|------|
| **过采样** | $f_s > 2f_{max}$ | 可完整重建，无失真 |
| **临界采样** | $f_s = 2f_{max}$ | 刚好可重建，理想情况 |
| **欠采样** | $f_s < 2f_{max}$ | 发生混叠，无法重建 |

### 2.4 混叠（Aliasing）现象

当采样频率不满足Nyquist条件时，会发生**混叠**：

**混叠原理**
- 采样后离散信号的频谱是原始连续信号频谱的周期延拓
- 延拓周期：$f_s$（采样频率）
- 当$f_s < 2f_{max}$时，相邻周期的频谱会发生重叠

**混叠的直观理解**
- 想象电影中快速旋转的车轮，看起来像是在缓慢倒转
- 这是因为采样率不足以捕捉车轮的真实旋转速度
- 在通信中，高频信号会被错误地识别为低频信号

**避免混叠的方法**
1. 提高采样率，使$f_s \geq 2f_{max}$
2. 在采样前使用**抗混叠滤波器**，将信号带宽限制在$f_s/2$以内

### 2.5 量化与编码

采样后的信号在时间上是离散的，但幅度仍然是连续的。**量化**将连续幅度映射到有限的离散值。

**量化过程**
- 量化级数$N$：决定数字信号能够表示的不同幅度值的数量
- 量化比特数$b = \log_2 N$：每个样本用多少比特表示
- 量化步长$\Delta = \frac{A_{max} - A_{min}}{2^b}$：相邻量化电平的间隔

**均匀量化 vs 非均匀量化**

| 类型 | 特点 | 应用场景 |
|------|------|----------|
| **均匀量化** | 量化步长固定 | 均匀分布信号 |
| **非均匀量化** | 对数尺度，大信号步长大 | 语音（压缩扩展） |

**A律/μ律压缩**是非均匀量化的典型例子，用于语音通信中。

**量化误差**
$$e[n] = x_q[n] - x[n]$$
其中$x_q[n]$是量化后的值，$x[n]$是原始值。

量化误差可以看作是一种噪声，称为**量化噪声**。

### 2.6 A/D转换完整流程

模拟信号到数字信号的转换过程：

```
模拟信号 → [抗混叠滤波器] → [采样器] → [量化器] → [编码器] → 数字信号
                ↓                ↓           ↓            ↓
           限制带宽         x[n]=x_c(nT)    离散幅度    二进制比特流
```

1. **抗混叠滤波器**：模拟低通滤波器，将信号限制在$f_s/2$以内
2. **采样器**：将连续信号在时域上离散化
3. **量化器**：将连续幅度映射到有限数量离散值
4. **编码器**：将量化值转换为二进制代码

### 3.3 量化误差分析

````python
# 量化误差分析
def demo_quantization():
    """
    演示均匀量化过程及其误差分析

    参数说明：
    - x_input: 输入连续信号
    - num_levels: 量化级数（2^b）
    """

    # 参数设置
    np.random.seed(42)

    # 生成测试信号（模拟语音或音频）
    N = 1000  # 样本数：足够大的样本量以获得有意义的统计结果
    x_input = np.random.randn(N) * 0.5 + np.sin(2 * np.pi * np.linspace(0, 1, N))  # 混合信号
    x_input = np.clip(x_input, -1, 1)  # 限制在[-1, 1]范围内

    def quantize(x, bits):
        """均匀量化函数"""
        num_levels = 2 ** bits
        delta = 2 / num_levels
        indices = np.floor((x + 1) / delta).astype(int)
        indices = np.clip(indices, 0, num_levels - 1)
        x_q = indices * delta - 1 + delta / 2
        return np.clip(x_q, -1, 1), delta, num_levels

    # 不同量化比特数
    bit_list = [2, 4, 8]

    fig, axes = plt.subplots(len(bit_list), 2, figsize=(14, 10))

    for i, bits in enumerate(bit_list):
        x_quantized, delta, num_levels = quantize(x_input, bits)

        # 量化误差
        error = x_quantized - x_input

        # 计算信噪比
        signal_power = np.mean(x_input ** 2)
        noise_power = np.mean(error ** 2)
        sqnr = 10 * np.log10(signal_power / noise_power)

        # 左图：原始信号与量化信号
        ax1 = axes[i, 0]
        ax1.plot(x_input[:200], 'b-', linewidth=1, alpha=0.5, label='原始信号')
        ax1.plot(x_quantized[:200], 'r-', linewidth=1.5, label=f'量化信号 ({bits}比特)')
        ax1.set_ylabel('幅度', fontsize=10)
        ax1.set_title(f'量化结果 (级数={num_levels}, Δ={delta:.3f})', fontsize=12)
        ax1.legend(loc='upper right')
        ax1.grid(True, alpha=0.3)
        ax1.set_ylim(-1.3, 1.3)

        # 右图：量化误差分布
        ax2 = axes[i, 1]
        ax2.hist(error, bins=50, density=True, alpha=0.7, color='orange', edgecolor='black')
        ax2.axvline(x=0, color='r', linestyle='--', linewidth=2)
        ax2.set_xlabel('误差', fontsize=10)
        ax2.set_ylabel('概率密度', fontsize=10)
        ax2.set_title(f'量化误差分布 (SQNR≈{sqnr:.1f}dB)', fontsize=12)
        ax2.grid(True, alpha=0.3)

    axes[0, 0].set_title('量化效果对比', fontsize=12)
    axes[0, 1].set_title('量化误差分布', fontsize=12)
    axes[-1, 0].set_xlabel('采样点', fontsize=10)
    axes[-1, 1].set_xlabel('误差值', fontsize=10)

    plt.tight_layout()
    plt.show()

    # 打印理论值与实际值对比
    print("量化比特数 | 量化级数 | 理论SQNR(dB) | 实际SQNR(dB)")
    print("------------|----------|--------------|------------")
    for bits in [2, 4, 8]:
        x_quantized, delta, num_levels = quantize(x_input, bits)
        error = x_quantized - x_input

        signal_power = np.mean(x_input ** 2)
        noise_power = np.mean(error ** 2)
        sqnr = 10 * np.log10(signal_power / noise_power)

        # 理论SQNR公式（近似）：SQNR(dB) ≈ 6.02 * b + 1.76
        theory_sqnr = 6.02 * bits + 1.76
        print(f"     {bits}      |    {num_levels:4d}    |    {theory_sqnr:6.2f}     |    {sqnr:6.2f}")

    return
````

````python
print("\n量化分析结论：")
print("1. 量化比特数越高，量化级数越多，量化误差越小")
print("2. 量化误差近似均匀分布，平均幅度约为±Δ/2")
print("3. 理论SQNR公式：SQNR(dB) ≈ 6.02 × b + 1.76 (b为量化比特数)")
print("4. 每增加1比特，SQNR约增加6dB（量化噪声功率减半）")
````

### 4.3 实际应用中的采样率选择

````python
# 实际应用中的采样率选择
def demo_sampling_rate_selection():
    """
    演示不同应用场景下的采样率选择原则
    """
    
    # 常见应用场景
    applications = [
        {'name': '语音信号', 'f_max': 3400, 'rate': 8000, 'standard': 'G.711 (μ-law/A-law)'},
        {'name': 'CD音频', 'f_max': 22000, 'rate': 44100, 'standard': 'CD-DA标准'},
        {'name': 'DVD音频', 'f_max': 22000, 'rate': 48000, 'standard': 'DVD-Audio'},
        {'name': '4G/5G基带', 'f_max': 10000000, 'rate': 30720000, 'standard': 'LTE/NR'},
        {'name': 'WiFi 802.11ac', 'f_max': 80000000, 'rate': 160000000, 'standard': '802.11ac'},
    ]
    
    print("常见应用场景的采样率选择：")
    print("=" * 80)
    print("{:<15} | {:>10} | {:>12} | {:>10} | {:<20}".format(
          "应用", "f_max(Hz)", "f_s(Hz)", "过采样率", "标准"))
    print("-" * 80)
    
    for app in applications:
        ratio = app['rate'] / (2 * app['f_max'])
        print("{:<15} | {:>10,d} | {:>12,d} | {:>10.1f}x | {:<20}".format(
              app['name'], app['f_max'], app['rate'], ratio, app['standard']))
    
    print("=" * 80)
    
    # 可视化
    fig, ax = plt.subplots(figsize=(12, 6))
    
    names = [app['name'] for app in applications]
    f_max_vals = [app['f_max'] for app in applications]
    rates = [app['rate'] for app in applications]
    
    x = np.arange(len(names))
    width = 0.35
    
    ax.bar(x - width/2, [f*2 for f in f_max_vals], width, label='Nyquist频率 (2f_max)', color='blue', alpha=0.7)
    ax.bar(x + width/2, rates, width, label='实际采样率', color='red', alpha=0.7)
    
    ax.set_xlabel('应用场景', fontsize=11)
    ax.set_ylabel('频率 (Hz)', fontsize=11)
    ax.set_title('不同应用场景的采样率对比', fontsize=12)
    ax.set_yscale('log')
    ax.set_xticks(x)
    ax.set_xticklabels(names, rotation=15, ha='right')
    ax.legend()
    ax.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.show()
    
    print("\n采样率设计原则：")
    print("1. 满足Nyquist定理：f_s ≥ 2f_max")
    print("2. 实际系统通常有过采样率（2-10倍），用于：")
    print("   - 放宽对抗混叠滤波器的要求")
    print("   - 提供设计裕量应对时钟偏差")
    print("   - 简化模拟前端设计")
    print("3. 特定应用有标准规定（如语音8kHz，CD 44.1kHz）")
    
demo_sampling_rate_selection()
````

## 总结

本notebook介绍了离散信号与采样的核心概念：

- **连续与离散信号**：连续信号在任意时刻有值，离散信号仅在特定采样点有定义
- **采样定理**：f_s ≥ 2f_max 是无失真重建的必要条件
- **混叠现象**：欠采样时高频成分折叠到低频，无法区分
- **抗混叠滤波**：在采样前滤除高于f_s/2的频率成分
- **量化编码**：将连续幅度映射到有限比特表示，引入量化误差
- **A/D转换流程**：滤波→采样→量化→编码的完整链路

这些概念是理解数字通信系统的基础，后续的调制、解调、滤波等处理都建立在离散信号处理之上。
