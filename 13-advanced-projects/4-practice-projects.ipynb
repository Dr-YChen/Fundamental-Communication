### 3.1 OTFS发送基类

下面是OTFS发送端的基类设计，提供了标准化的接口和可扩展的框架。

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

print("OTFS发送接收基类框架加载成功")
````

````python
class OTFSTransmitterBase:
    """
    OTFS发送端基类
    
    提供标准化的OTFS发送端接口，子类可继承并扩展特定功能
    
    使用方法:
        transmitter = OTFSTransmitterBase(N=32, M=32, qam_order=16)
        tx_signal = transmitter.transmit(bits)
    """
    
    def __init__(self, N=32, M=32, qam_order=16, cp_len=8):
        """
        初始化OTFS发送端
        
        参数:
            N: 时间维度（多普勒采样数）
            M: 频率维度（延迟采样数）
            qam_order: QAM调制阶数（4, 16, 64, 256）
            cp_len: 循环前缀长度
        """
        self.N = N
        self.M = M
        self.qam_order = qam_order
        self.cp_len = cp_len
        
        # 创建星座图
        self.constellation = self._create_qam_constellation()
        self.bits_per_symbol = int(np.log2(qam_order))
        
        print(f"OTFS发送端初始化: N={N}, M={M}, QAM={qam_order}")
        print(f"网格大小: {N}x{M} = {N*M}符号, CP长度: {cp_len}")
    
    def _create_qam_constellation(self):
        """
        创建QAM星座图
        
        返回:
            constellation: QAM星座点数组
        """
        if self.qam_order == 4:
            return np.array([1+0j, 0+1j, -1+0j, 0-1j])
        elif self.qam_order == 16:
            levels = np.array([-3, -1, 1, 3]) / 3.0
            constellation = []
            for i in levels:
                for j in levels:
                    constellation.append(i + 1j * j)
            return np.array(constellation)
        elif self.qam_order == 64:
            levels = np.array([-7, -5, -3, -1, 1, 3, 5, 7]) / 7.0
            constellation = []
            for i in levels:
                for j in levels:
                    constellation.append(i + 1j * j)
            return np.array(constellation)
        else:
            raise ValueError(f"不支持的QAM阶数: {self.qam_order}")
    
    def qam_modulate(self, bits):
        """
        将比特流映射到QAM星座点
        
        参数:
            bits: 输入比特流 (numpy array)
        
        返回:
            symbols: QAM符号数组
        """
        symbols = []
        for i in range(0, len(bits), self.bits_per_symbol):
            if i + self.bits_per_symbol <= len(bits):
                symbol_bits = bits[i:i+self.bits_per_symbol]
                symbol_idx = int(''.join(map(str, symbol_bits)), 2)
                symbols.append(self.constellation[symbol_idx])
        return np.array(symbols)
    
    def place_in_dd_grid(self, symbols):
        """
        将QAM符号放置到延迟-多普勒网格
        
        参数:
            symbols: QAM符号数组，长度应为N*M
        
        返回:
            X_dd: 延迟-多普勒域信号，shape (M, N)
        """
        if len(symbols) != self.N * self.M:
            raise ValueError(f"符号数量 {len(symbols)} 与网格大小 {self.N*self.M} 不匹配")
        return symbols.reshape(self.M, self.N)
    
    def sfft(self, x_dd):
        """
        对称有限傅里叶变换（SFFT）
        
        将延迟-多普勒域信号变换到时频域
        
        参数:
            x_dd: 延迟-多普勒域信号，shape (M, N)
        
        返回:
            X_tf: 时频域信号，shape (N, M)
        """
        # 步骤1：沿列方向FFT（多普勒→时间）
        X_temp = np.fft.fft(x_dd, axis=0)
        # 步骤2：沿行方向IFFT（延迟→频率）
        X_tf = np.fft.ifft(X_temp, axis=1)
        return X_tf
    
    def ofdm_modulate(self, tf_grid):
        """
        OFDM调制：将时频域信号转换为时域信号
        
        参数:
            tf_grid: 时频域信号，shape (N, M)
        
        返回:
            ofdm_symbols: 时域OFDM符号，shape (N, M+cp_len)
        """
        ofdm_symbols = []
        for i in range(self.N):
            # IFFT：频域→时域
            time_domain = np.fft.ifft(tf_grid[i], n=self.M)
            # 添加循环前缀
            cp = time_domain[-self.cp_len:]
            ofdm_with_cp = np.concatenate([cp, time_domain])
            ofdm_symbols.append(ofdm_with_cp)
        return np.array(ofdm_symbols)
    
    def add_pilot(self, tx_signal, pilot_pattern='center'):
        """
        添加导频信号（可被子类重写）
        
        参数:
            tx_signal: 发送信号
            pilot_pattern: 导频图案 ('center', 'distributed')
        
        返回:
            带导频的信号
        """
        # 默认实现：不做任何处理
        return tx_signal
    
    def transmit(self, bits):
        """
        完整发送流程
        
        参数:
            bits: 输入比特流
        
        返回:
            tx_signal: 发送信号（含CP）
        """
        # 1. QAM调制
        symbols = self.qam_modulate(bits)
        
        # 2. 放置到DD网格
        X_dd = self.place_in_dd_grid(symbols)
        
        # 3. SFFT变换
        X_tf = self.sfft(X_dd)
        
        # 4. OFDM调制
        tx_signal = self.ofdm_modulate(X_tf)
        
        # 5. 添加导频（可选）
        tx_signal = self.add_pilot(tx_signal)
        
        return tx_signal
    
    def get_info(self):
        """
        获取发送端配置信息
        """
        return {
            'N': self.N,
            'M': self.M,
            'qam_order': self.qam_order,
            'cp_len': self.cp_len,
            'bits_per_symbol': self.bits_per_symbol,
            'total_symbols': self.N * self.M
        }
````

### 3.2 OTFS接收基类

````python
class OTFSReceiverBase:
    """
    OTFS接收端基类
    
    提供标准化的OTFS接收端接口
    
    使用方法:
        receiver = OTFSReceiverBase(N=32, M=32, qam_order=16)
        rx_bits = receiver.receive(rx_signal, channel_estimate)
    """
    
    def __init__(self, N=32, M=32, qam_order=16, cp_len=8):
        """
        初始化OTFS接收端
        
        参数:
            N: 时间维度（多普勒采样数）
            M: 频率维度（延迟采样数）
            qam_order: QAM调制阶数
            cp_len: 循环前缀长度
        """
        self.N = N
        self.M = M
        self.qam_order = qam_order
        self.cp_len = cp_len
        
        # 创建星座图
        self.constellation = self._create_qam_constellation()
        self.bits_per_symbol = int(np.log2(qam_order))
        
        print(f"OTFS接收端初始化: N={N}, M={M}, QAM={qam_order}")
    
    def _create_qam_constellation(self):
        """
        创建QAM星座图（与发送端相同）
        """
        if self.qam_order == 4:
            return np.array([1+0j, 0+1j, -1+0j, 0-1j])
        elif self.qam_order == 16:
            levels = np.array([-3, -1, 1, 3]) / 3.0
            constellation = []
            for i in levels:
                for j in levels:
                    constellation.append(i + 1j * j)
            return np.array(constellation)
        elif self.qam_order == 64:
            levels = np.array([-7, -5, -3, -1, 1, 3, 5, 7]) / 7.0
            constellation = []
            for i in levels:
                for j in levels:
                    constellation.append(i + 1j * j)
            return np.array(constellation)
        else:
            raise ValueError(f"不支持的QAM阶数: {self.qam_order}")
    
    def ofdm_demodulate(self, rx_ofdm_symbols):
        """
        OFDM解调：将时域信号转换为频域
        
        参数:
            rx_ofdm_symbols: 接收的OFDM符号，shape (N, M+cp_len)
        
        返回:
            Y_tf: 时频域信号，shape (N, M)
        """
        tf_grid = []
        for i in range(self.N):
            # 去除CP
            symbol = rx_ofdm_symbols[i, self.cp_len:self.cp_len + self.M]
            # FFT：时域→频域
            freq_domain = np.fft.fft(symbol, n=self.M)
            tf_grid.append(freq_domain)
        return np.array(tf_grid)
    
    def isfft(self, X_tf):
        """
        逆对称有限傅里叶变换（ISFFT）
        
        将时频域信号变换到延迟-多普勒域
        
        参数:
            X_tf: 时频域信号，shape (N, M)
        
        返回:
            Y_dd: 延迟-多普勒域信号，shape (M, N)
        """
        # 逆变换：步骤1 FFT列，步骤2 IFFT行
        y_temp = np.fft.fft(X_tf, axis=0)
        Y_dd = np.fft.ifft(y_temp, axis=1)
        return Y_dd
    
    def equalize(self, Y_dd, channel_estimate, method='zf'):
        """
        信道均衡（可被子类重写）
        
        参数:
            Y_dd: 接收的DD域信号
            channel_estimate: 信道估计结果
            method: 均衡方法 ('zf', 'mmse')
        
        返回:
            Y_equalized: 均衡后的信号
        """
        # 单抽头均衡：简化实现
        if channel_estimate is None:
            return Y_dd
        
        # 假设信道在主径处估计
        h_main = channel_estimate.get('main_tap', 1.0 + 0j)
        
        if method == 'zf':
            return Y_dd / h_main
        elif method == 'mmse':
            snr = channel_estimate.get('snr', 100)
            h_conj = np.conj(h_main)
            h_abs_sq = np.abs(h_main) ** 2
            Y_equalized = h_conj / (h_abs_sq + 1/snr) * Y_dd
            return Y_equalized
        else:
            return Y_dd / h_main
    
    def qam_demodulate(self, symbols):
        """
        将QAM符号解映射为比特流
        
        参数:
            symbols: QAM符号数组
        
        返回:
            bits: 解调后的比特流
        """
        recovered_bits = []
        for symbol in symbols:
            # 找到最近的星座点
            distances = np.abs(self.constellation - symbol)
            nearest_idx = np.argmin(distances)
            # 转换为比特
            bit_str = format(nearest_idx, f'0{self.bits_per_symbol}b')
            recovered_bits.extend([int(b) for b in bit_str])
        return np.array(recovered_bits)
    
    def receive(self, rx_signal, channel_estimate=None, equalization_method='zf'):
        """
        完整接收流程
        
        参数:
            rx_signal: 接收信号（含CP）
            channel_estimate: 信道估计结果
            equalization_method: 均衡方法
        
        返回:
            rx_bits: 解调后的比特流
        """
        # 1. OFDM解调
        Y_tf = self.ofdm_demodulate(rx_signal)
        
        # 2. ISFFT变换
        Y_dd = self.isfft(Y_tf)
        
        # 3. 信道均衡
        Y_equalized = self.equalize(Y_dd, channel_estimate, equalization_method)
        
        # 4. 从网格提取符号
        received_symbols = Y_equalized.flatten()
        
        # 5. QAM解映射
        rx_bits = self.qam_demodulate(received_symbols)
        
        return rx_bits
````

### 3.3 OTFS系统基类（发送+接收）

````python
class OTFSCommunicationSystem:
    """
    OTFS通信系统基类
    
    整合发送端和接收端，提供完整的通信系统接口
    
    使用方法:
        system = OTFSCommunicationSystem(N=32, M=32)
        ber = system.run(tx_bits, snr_db=20)
    """
    
    def __init__(self, N=32, M=32, qam_order=16, cp_len=8):
        """
        初始化OTFS通信系统
        """
        self.N = N
        self.M = M
        self.qam_order = qam_order
        self.cp_len = cp_len
        
        # 创建发送端和接收端
        self.transmitter = OTFSTransmitterBase(N, M, qam_order, cp_len)
        self.receiver = OTFSReceiverBase(N, M, qam_order, cp_len)
        
        print(f"\nOTFS通信系统初始化完成")
        print(f"总带宽: {N*M} 符号/帧")
    
    def create_channel(self, num_paths=3, max_delay=5, max_doppler=3, seed=42):
        """
        创建延迟-多普勒域信道
        
        参数:
            num_paths: 多径数量
            max_delay: 最大延迟索引
            max_doppler: 最大多普勒偏移
            seed: 随机种子
        
        返回:
            channel: 信道冲激响应字典
        """
        np.random.seed(seed)
        
        # 主径
        channel = {
            'taps': [(0, 0, 1.0 + 0j)],  # (delay, doppler, amplitude)
            'max_delay': max_delay,
            'max_doppler': max_doppler
        }
        
        # 添加随机路径
        for _ in range(num_paths - 1):
            delay = np.random.randint(1, max_delay + 1)
            doppler = np.random.randint(-max_doppler, max_doppler + 1)
            amp = (np.random.randn() + 1j * np.random.randn()) * 0.3
            channel['taps'].append((delay, doppler, amp))
        
        return channel
    
    def apply_channel(self, tx_signal, channel):
        """
        将信道应用于发送信号
        
        参数:
            tx_signal: 发送信号
            channel: 信道参数
        
        返回:
            rx_signal: 接收信号
        """
        rx_signal = np.zeros_like(tx_signal, dtype=complex)
        
        for delay, doppler, amp in channel['taps']:
            for sym_idx in range(self.N):
                # 多普勒相位旋转
                doppler_phase = np.exp(1j * 2 * np.pi * doppler * sym_idx / self.N)
                gain = amp * doppler_phase
                
                # 应用延迟
                start = self.cp_len + delay
                end = start + self.M
                if end <= tx_signal.shape[1]:
                    rx_signal[sym_idx, start:end] += tx_signal[sym_idx, self.cp_len:self.cp_len+self.M] * gain
        
        return rx_signal
    
    def add_noise(self, signal, snr_db):
        """
        添加高斯噪声
        
        参数:
            signal: 输入信号
            snr_db: 信噪比(dB)
        
        返回:
            noisy_signal: 加噪后的信号
        """
        signal_power = np.mean(np.abs(signal) ** 2)
        noise_power = signal_power / (10 ** (snr_db / 10))
        noise = np.sqrt(noise_power / 2) * (
            np.random.randn(*signal.shape) + 1j * np.random.randn(*signal.shape)
        )
        return signal + noise
    
    def estimate_channel(self, channel):
        """
        估计信道（简化版本：直接返回真实信道）
        
        参数:
            channel: 真实信道
        
        返回:
            channel_estimate: 信道估计结果
        """
        # 简化：假设完美信道估计
        return {'main_tap': 1.0 + 0j, 'snr': 100}
    
    def run(self, tx_bits, snr_db=20, channel=None, equalization_method='zf'):
        """
        运行完整通信系统
        
        参数:
            tx_bits: 发送比特流
            snr_db: 信噪比(dB)
            channel: 信道参数（可选）
            equalization_method: 均衡方法
        
        返回:
            result: 包含BER和详细信息的字典
        """
        # 发送
        tx_signal = self.transmitter.transmit(tx_bits)
        
        # 创建或使用信道
        if channel is None:
            channel = self.create_channel()
        
        # 通过信道
        rx_signal = self.apply_channel(tx_signal, channel)
        
        # 添加噪声
        rx_signal = self.add_noise(rx_signal, snr_db)
        
        # 信道估计
        channel_estimate = self.estimate_channel(channel)
        channel_estimate['snr'] = 10 ** (snr_db / 10)
        
        # 接收
        rx_bits = self.receiver.receive(rx_signal, channel_estimate, equalization_method)
        
        # 计算BER
        min_len = min(len(tx_bits), len(rx_bits))
        num_errors = np.sum(tx_bits[:min_len] != rx_bits[:min_len])
        ber = num_errors / min_len if min_len > 0 else 0
        
        return {
            'ber': ber,
            'num_errors': num_errors,
            'num_bits': min_len,
            'snr_db': snr_db,
            'channel': channel
        }
    
    def run_ber_sweep(self, tx_bits, snr_range, **kwargs):
        """
        扫描不同SNR下的BER性能
        
        参数:
            tx_bits: 发送比特流
            snr_range: SNR范围 (dB)
            **kwargs: 其他参数传递给run
        
        返回:
            results: BER结果列表
        """
        results = []
        for snr in snr_range:
            res = self.run(tx_bits, snr_db=snr, **kwargs)
            results.append(res)
        return results
````

### 3.4 扩展：MIMO-OTFS发送接收基类

````python
class MIMOOTFSCommunicationSystem:
    """
    MIMO-OTFS通信系统基类
    
    支持多天线配置的OTFS通信系统
    
    使用方法:
        system = MIMOOTFSCommunicationSystem(N_t=2, N_r=2, N=32, M=32)
        ber = system.run(tx_bits_list, snr_db=20)
    """
    
    def __init__(self, N_t=2, N_r=2, N=32, M=32, qam_order=16, cp_len=8):
        """
        初始化MIMO-OTFS通信系统
        
        参数:
            N_t: 发射天线数
            N_r: 接收天线数
            N: 时间维度
            M: 频率维度
            qam_order: QAM调制阶数
            cp_len: 循环前缀长度
        """
        self.N_t = N_t
        self.N_r = N_r
        self.N = N
        self.M = M
        self.qam_order = qam_order
        self.cp_len = cp_len
        
        # 创建星座
        self.constellation = self._create_qam_constellation()
        self.bits_per_symbol = int(np.log2(qam_order))
        
        print(f"MIMO-OTFS系统初始化: {N_t}x{N_r} MIMO")
        print(f"OTFS网格: {N}x{M}, 调制: {qam_order}-QAM")
    
    def _create_qam_constellation(self):
        """创建QAM星座图"""
        if self.qam_order == 16:
            levels = np.array([-3, -1, 1, 3]) / 3.0
            constellation = []
            for i in levels:
                for j in levels:
                    constellation.append(i + 1j * j)
            return np.array(constellation)
        return np.array([1+0j, 0+1j, -1+0j, 0-1j])
    
    def sfft(self, x_dd):
        """SFFT变换"""
        X_temp = np.fft.fft(x_dd, axis=0)
        return np.fft.ifft(X_temp, axis=1)
    
    def isfft(self, X_tf):
        """ISFFT变换"""
        y_temp = np.fft.fft(X_tf, axis=0)
        return np.fft.ifft(y_temp, axis=1)
    
    def ofdm_modulate(self, tf_grid):
        """OFDM调制"""
        ofdm_symbols = []
        for i in range(self.N):
            time_domain = np.fft.ifft(tf_grid[i], n=self.M)
            cp = time_domain[-self.cp_len:]
            ofdm_with_cp = np.concatenate([cp, time_domain])
            ofdm_symbols.append(ofdm_with_cp)
        return np.array(ofdm_symbols)
    
    def ofdm_demodulate(self, rx_ofdm_symbols):
        """OFDM解调"""
        tf_grid = []
        for i in range(self.N):
            symbol = rx_ofdm_symbols[i, self.cp_len:self.cp_len + self.M]
            freq_domain = np.fft.fft(symbol, n=self.M)
            tf_grid.append(freq_domain)
        return np.array(tf_grid)
    
    def create_mimo_channel(self, num_paths=3, max_delay=5, max_doppler=3, seed=42):
        """
        创建MIMO信道
        
        返回:
            H_dd: shape (N_r, N_t, M, N) 信道张量
        """
        np.random.seed(seed)
        H_dd = np.zeros((self.N_r, self.N_t, self.M, self.N), dtype=complex)
        
        for n_t in range(self.N_t):
            for n_r in range(self.N_r):
                # 主径
                H_dd[n_r, n_t, 0, max_doppler] = 1.0 + 0j
                
                # 随机路径
                for _ in range(num_paths - 1):
                    delay_idx = np.random.randint(1, max_delay + 1)
                    doppler_idx = np.random.randint(0, 2 * max_doppler + 1) - max_doppler
                    amp = (np.random.randn() + 1j * np.random.randn()) * 0.3
                    
                    if delay_idx < self.M and (doppler_idx + max_doppler) < self.N:
                        H_dd[n_r, n_t, delay_idx, doppler_idx + max_doppler] = amp
        
        return H_dd
    
    def transmit(self, tx_bits_list):
        """
        MIMO发送
        
        参数:
            tx_bits_list: N_t个比特流列表
        
        返回:
            tx_ofdm_list: N_t个发送信号列表
        """
        tx_ofdm_list = []
        
        for nt in range(self.N_t):
            # QAM调制
            bits = tx_bits_list[nt]
            symbols = []
            for i in range(0, len(bits), self.bits_per_symbol):
                if i + self.bits_per_symbol <= len(bits):
                    symbol_bits = bits[i:i+self.bits_per_symbol]
                    symbol_idx = int(''.join(map(str, symbol_bits)), 2)
                    symbols.append(self.constellation[symbol_idx])
            
            # DD域放置
            X_dd = np.array(symbols).reshape(self.M, self.N)
            
            # SFFT
            X_tf = self.sfft(X_dd)
            
            # OFDM调制
            tx_ofdm = self.ofdm_modulate(X_tf)
            tx_ofdm_list.append(tx_ofdm)
        
        return tx_ofdm_list
    
    def receive(self, rx_ofdm_list, H_dd, snr_db, detector_type='mmse'):
        """
        MIMO接收
        
        参数:
            rx_ofdm_list: N_r个接收信号列表
            H_dd: MIMO信道张量
            snr_db: 信噪比(dB)
            detector_type: 检测器类型
        
        返回:
            rx_bits_list: N_t个恢复的比特流列表
        """
        # OFDM解调 + ISFFT
        Y_dd_list = []
        for nr in range(self.N_r):
            Y_tf = self.ofdm_demodulate(rx_ofdm_list[nr])
            Y_dd = self.isfft(Y_tf)
            Y_dd_list.append(Y_dd)
        
        # 在主径处进行检测
        max_doppler = 3
        H_main = H_dd[:, :, 0, max_doppler]
        
        # 收集每个OFDM符号的检测结果
        detected_symbols_all = []
        
        for sym_idx in range(self.N):
            # 收集该符号时刻的接收信号
            y = np.array([Y_dd_list[nr][0, max_doppler] for nr in range(self.N_r)])
            
            # 检测
            if detector_type == 'zf':
                x_hat = np.linalg.pinv(H_main) @ y
            elif detector_type == 'mmse':
                SNR = 10 ** (snr_db / 10)
                sigma2 = 1 / SNR
                x_hat = H_main.conj().T @ np.linalg.inv(H_main @ H_main.conj().T + sigma2 * np.eye(self.N_r)) @ y
            else:
                x_hat = np.linalg.pinv(H_main) @ y
            
            detected_symbols_all.extend(x_hat)
        
        # QAM解映射
        recovered_bits = []
        for symbol in detected_symbols_all:
            distances = np.abs(self.constellation - symbol)
            nearest_idx = np.argmin(distances)
            bit_str = format(nearest_idx, f'0{self.bits_per_symbol}b')
            recovered_bits.extend([int(b) for b in bit_str])
        
        return np.array(recovered_bits)
    
    def apply_channel(self, tx_ofdm_list, H_dd):
        """应用MIMO信道"""
        rx_ofdm_list = [np.zeros_like(tx_ofdm_list[0], dtype=complex) for _ in range(self.N_r)]
        
        max_delay = 5
        max_doppler = 3
        
        for n_t in range(self.N_t):
            for n_r in range(self.N_r):
                for delay_idx in range(self.M):
                    for doppler_idx in range(self.N):
                        amp = H_dd[n_r, n_t, delay_idx, doppler_idx]
                        if np.abs(amp) > 1e-6:
                            doppler_phase = np.exp(1j * 2 * np.pi * (doppler_idx - max_doppler) * np.arange(self.N) / self.N)
                            gain = amp * doppler_phase
                            
                            start = self.cp_len + delay_idx
                            end = start + self.M
                            if end <= tx_ofdm_list[0].shape[1]:
                                rx_ofdm_list[n_r][:, start:end] += tx_ofdm_list[n_t][:, self.cp_len:self.cp_len+self.M] * gain[:, np.newaxis]
        
        return rx_ofdm_list
    
    def run(self, tx_bits_list, snr_db=20, detector_type='mmse'):
        """
        运行MIMO-OTFS系统
        
        参数:
            tx_bits_list: N_t个比特流列表
            snr_db: 信噪比(dB)
            detector_type: 检测器类型
        
        返回:
            result: 包含BER等信息的字典
        """
        # 创建信道
        H_dd = self.create_mimo_channel()
        
        # 发送
        tx_ofdm_list = self.transmit(tx_bits_list)
        
        # 通过信道
        rx_ofdm_list = self.apply_channel(tx_ofdm_list, H_dd)
        
        # 添加噪声
        signal_power = sum(np.mean(np.abs(rx)**2) for rx in rx_ofdm_list) / self.N_r
        noise_power = signal_power / (10**(snr_db/10))
        for nr in range(self.N_r):
            noise = np.sqrt(noise_power/2) * (np.random.randn(*rx_ofdm_list[nr].shape) + 
                                               1j * np.random.randn(*rx_ofdm_list[nr].shape))
            rx_ofdm_list[nr] += noise
        
        # 接收和检测
        rx_bits = self.receive(rx_ofdm_list, H_dd, snr_db, detector_type)
        
        # 合并发射比特并计算BER
        tx_bits_combined = np.concatenate(tx_bits_list)
        min_len = min(len(tx_bits_combined), len(rx_bits))
        num_errors = np.sum(tx_bits_combined[:min_len] != rx_bits[:min_len])
        ber = num_errors / min_len if min_len > 0 else 0
        
        return {'ber': ber, 'num_errors': num_errors, 'num_bits': min_len}
````

## 4. 项目评估标准

### 4.1 功能完整性评分

````python
def evaluate_functionality(project_type, requirements):
    """
    评估功能完整性
    
    参数:
        project_type: 项目类型 ('basic', 'intermediate', 'advanced', 'expert')
        requirements: 需求清单字典
        
    返回:
        score: 功能完整性得分 (0-100)
        details: 详细评分
    """
    scores = {}
    total_weight = 0
    
    # 基础功能权重
    weights = {
        'modulation': 20,
        'demodulation': 20,
        'channel_estimation': 15,
        'equalization': 15,
        'error_correction': 10,
        'synchronization': 10,
        'performance_metrics': 10
    }
    
    for req, achieved in requirements.items():
        if req in weights:
            weight = weights[req]
            total_weight += weight
            scores[req] = achieved * weight
    
    total_score = sum(scores.values()) / total_weight * 100 if total_weight > 0 else 0
    
    return total_score, scores

print("功能完整性评估框架已定义")
````

### 4.2 性能指标评分

````python
class PerformanceEvaluator:
    """
    OTFS系统性能评估器
    
    提供标准化的性能评估方法和指标计算
    """
    
    def __init__(self):
        self.metrics = {}
    
    def evaluate_ber_performance(self, ber_results, snr_range, target_ber=1e-4):
        """
        评估BER性能
        
        参数:
            ber_results: BER结果列表
            snr_range: 对应的SNR范围
            target_ber: 目标BER
        
        返回:
            score: BER性能得分 (0-100)
            details: 详细分析
        """
        # 找到满足目标BER的最小SNR
        min_snr_target = None
        for snr, ber in zip(snr_range, ber_results):
            if ber <= target_ber:
                min_snr_target = snr
                break
        
        # 计算BER曲线下面积（越小越好）
        from scipy import integrate
        
        # 将BER转换为log域
        log_ber = np.log10(np.maximum(ber_results, 1e-10))
        snr_linear = np.array(snr_range)
        
        try:
            auc = integrate.trapz(-log_ber, snr_linear)
        except:
            auc = np.sum(-log_ber) * (snr_range[1] - snr_range[0])
        
        # 评分：SNR阈值和AUC综合评分
        if min_snr_target is None:
            snr_score = 0
        else:
            # SNR越小越好，30dB为满分
            snr_score = max(0, min(50, (30 - min_snr_target) * 2))
        
        # AUC评分（满分50）
        auc_score = min(50, auc / 100)
        
        total_score = snr_score + auc_score
        
        return total_score, {
            'min_snr_target': min_snr_target,
            'snr_score': snr_score,
            'auc_score': auc_score,
            'auc': auc
        }
    
    def evaluate_spectral_efficiency(self, system_config, ber_results, target_ber=1e-4):
        """
        评估频谱效率
        
        参数:
            system_config: 系统配置字典
            ber_results: BER结果
        
        返回:
            score: 频谱效率得分
            spectral_efficiency: 实际频谱效率 bps/Hz
        """
        N = system_config.get('N', 32)
        M = system_config.get('M', 32)
        qam_order = system_config.get('qam_order', 16)
        
        # 理论频谱效率
        bits_per_symbol = np.log2(qam_order)
        spectral_eff = bits_per_symbol  # 简化计算（未考虑CP等开销）
        
        # 实际效率（根据BER调整）
        avg_ber = np.mean(ber_results)
        if avg_ber > 0.1:
            actual_eff = spectral_eff * 0.5
        elif avg_ber > 0.01:
            actual_eff = spectral_eff * 0.8
        else:
            actual_eff = spectral_eff * 0.95
        
        # 评分（满分100）
        score = (actual_eff / spectral_eff) * 100
        
        return score, actual_eff
    
    def evaluate_complexity(self, execution_time, memory_usage, target_time=None):
        """
        评估算法复杂度
        
        参数:
            execution_time: 执行时间（秒）
            memory_usage: 内存使用量（MB）
            target_time: 目标执行时间
        
        返回:
            score: 复杂度得分 (0-100)
        """
        if target_time is None:
            target_time = 1.0  # 默认1秒
        
        # 时间评分
        if execution_time <= target_time:
            time_score = 50
        else:
            time_score = max(0, 50 * target_time / execution_time)
        
        # 内存评分（假设100MB为满分）
        if memory_usage <= 100:
            mem_score = 50
        else:
            mem_score = max(0, 50 * 100 / memory_usage)
        
        return time_score + mem_score, {'time_score': time_score, 'mem_score': mem_score}
    
    def generate_report(self, all_metrics):
        """
        生成综合评估报告
        
        参数:
            all_metrics: 所有评估指标
        
        返回:
            report: 评估报告字典
        """
        total_score = sum(all_metrics.values())
        
        report = {
            'total_score': total_score,
            'grade': self._get_grade(total_score),
            'details': all_metrics,
            'recommendations': self._get_recommendations(all_metrics)
        }
        
        return report
    
    def _get_grade(self, score):
        """根据得分确定等级"""
        if score >= 90:
            return 'A'
        elif score >= 80:
            return 'B'
        elif score >= 70:
            return 'C'
        elif score >= 60:
            return 'D'
        else:
            return 'F'
    
    def _get_recommendations(self, metrics):
        """根据评估结果给出建议"""
        recs = []
        if metrics.get('ber_score', 0) < 60:
            recs.append("BER性能有待提升，建议优化均衡算法或增加信道编码")
        if metrics.get('efficiency_score', 0) < 60:
            recs.append("频谱效率较低，可考虑使用更高阶调制")
        if metrics.get('complexity_score', 0) < 60:
            recs.append("算法复杂度较高，建议优化实现或使用近似算法")
        return recs

print("性能评估器已定义")
````

````python
# 演示性能评估
evaluator = PerformanceEvaluator()

# 假设有BER测试结果
snr_range = [5, 10, 15, 20, 25, 30]
ber_results = [0.1, 0.05, 0.01, 0.001, 0.0001, 0.00001]

ber_score, ber_details = evaluator.evaluate_ber_performance(ber_results, snr_range, target_ber=1e-4)
print(f"BER性能得分: {ber_score:.2f}/100")
print(f"  达标最小SNR: {ber_details['min_snr_target']} dB")
print(f"  SNR评分: {ber_details['snr_score']:.2f}")
print(f"  AUC评分: {ber_details['auc_score']:.2f}")
````

### 4.3 代码质量评分

````python
class CodeQualityEvaluator:
    """
    代码质量评估器
    
    评估OTFS实现代码的质量
    """
    
    def __init__(self):
        self.weights = {
            'readability': 20,
            'modularity': 20,
            'documentation': 20,
            'testability': 20,
            'performance': 20
        }
    
    def evaluate_readability(self, code_structure):
        """
        评估代码可读性
        
        检查：
        - 变量命名规范
        - 函数长度适中
        - 缩进和格式
        - 注释清晰度
        """
        score = 80  # 默认基础分
        details = {}
        
        # 检查注释覆盖率
        comment_ratio = code_structure.get('comment_lines', 0) / max(code_structure.get('total_lines', 1), 1)
        if comment_ratio > 0.2:
            score += 10
        elif comment_ratio < 0.1:
            score -= 20
        
        details['comment_ratio'] = comment_ratio
        details['score'] = min(100, max(0, score))
        
        return min(100, max(0, score)), details
    
    def evaluate_modularity(self, class_count, function_count, dependencies):
        """
        评估模块化程度
        
        检查：
        - 类和函数数量
        - 依赖复杂度
        - 耦合度
        """
        score = 70  # 默认基础分
        
        # 模块数量评分
        if class_count + function_count > 5:
            score += 10
        if class_count + function_count > 10:
            score += 10
        
        # 依赖评分
        if len(dependencies) < 5:
            score += 10
        
        details = {
            'class_count': class_count,
            'function_count': function_count,
            'dependency_count': len(dependencies)
        }
        
        return min(100, score), details
    
    def evaluate_documentation(self, has_docstrings, has_comments, has_readme):
        """
        评估文档完整性
        """
        score = 0
        
        if has_docstrings:
            score += 40
        if has_comments:
            score += 30
        if has_readme:
            score += 30
        
        return score, {
            'has_docstrings': has_docstrings,
            'has_comments': has_comments,
            'has_readme': has_readme
        }
    
    def evaluate_testability(self, has_tests, test_coverage):
        """
        评估可测试性
        """
        score = 0
        
        if has_tests:
            score += 50
        if test_coverage > 0.5:
            score += 30
        if test_coverage > 0.8:
            score += 20
        
        return score, {
            'has_tests': has_tests,
            'test_coverage': test_coverage
        }
    
    def evaluate_performance(self, execution_time, memory_usage):
        """
        评估性能优化程度
        """
        score = 100
        
        # 检查是否使用了向量化操作
        # 检查是否有不必要的循环
        # 检查内存使用是否合理
        
        # 时间惩罚
        if execution_time > 10:
            score -= 20
        if execution_time > 60:
            score -= 30
        
        # 内存惩罚
        if memory_usage > 500:
            score -= 20
        if memory_usage > 1000:
            score -= 30
        
        return max(0, score), {
            'execution_time': execution_time,
            'memory_usage': memory_usage
        }
    
    def generate_report(self, code_structure, class_count, function_count, 
                       dependencies, has_tests, test_coverage,
                       execution_time, memory_usage):
        """
        生成代码质量评估报告
        """
        readability_score, r_details = self.evaluate_readability(code_structure)
        modularity_score, m_details = self.evaluate_modularity(class_count, function_count, dependencies)
        doc_score, d_details = self.evaluate_documentation(True, True, True)
        test_score, t_details = self.evaluate_testability(has_tests, test_coverage)
        perf_score, p_details = self.evaluate_performance(execution_time, memory_usage)
        
        scores = {
            'readability': readability_score,
            'modularity': modularity_score,
            'documentation': doc_score,
            'testability': test_score,
            'performance': perf_score
        }
        
        weights = self.weights
        total_weight = sum(weights.values())
        total_score = sum(scores[k] * weights[k] / 100 for k in weights.keys()) / total_weight * 100
        
        return {
            'total_score': total_score,
            'scores': scores,
            'details': {
                'readability': r_details,
                'modularity': m_details,
                'documentation': d_details,
                'testability': t_details,
                'performance': p_details
            }
        }

print("代码质量评估器已定义")
````
